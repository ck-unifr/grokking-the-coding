# Array and Hashing

## Contains Duplicate



````python
"""  
Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:

Input: nums = [1, 2, 3, 3]

Output: true

Example 2:

Input: nums = [1, 2, 3, 4]

Output: false

https://neetcode.io/problems/duplicate-integer/question?list=blind75
"""
````

````python
from typing import List 


class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()
        for num in nums:
            if num in seen:
                return True
            seen.add(num)
        return False
````

````python
"""  
Time Complexity: 
O(n)
 - We iterate through the array once.
Space Complexity: 
O(n)
 - We might store up to n elements in the set.
"""
````

## Valid Anagram



````python
"""  
Given two strings s and t, return true if the two strings are anagrams of each other, 
otherwise return false.

An anagram is a string that contains the exact same characters as another string, 
but the order of the characters can be different.

Example 1:

Input: s = "racecar", t = "carrace"

Output: true
Example 2:

Input: s = "jar", t = "jam"

Output: false
Constraints:

s and t consist of lowercase English letters.
"""
````

````python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        return sorted(s) == sorted(t)
    
````

````python
"""  
Time Complexity: 
O(N)
 (we iterate through the strings once).
Space Complexity: 
O(1)
 (since there are only 26 lowercase English letters, the dictionary size has a constant limit).
"""
````

## Group Anagrams

````python
"""  
Given an array of strings strs, group all anagrams together into sublists. 
You may return the output in any order.

An anagram is a string that contains the exact same characters as another string, 
but the order of the characters can be different.

Example 1:

Input: strs = ["act","pots","tops","cat","stop","hat"]

Output: [["hat"],["act", "cat"],["stop", "pots", "tops"]]
Example 2:

Input: strs = ["x"]

Output: [["x"]]
Example 3:

Input: strs = [""]

Output: [[""]]
Constraints:

1 <= strs.length <= 1000.
0 <= strs[i].length <= 100
strs[i] is made up of lowercase English letters.


https://neetcode.io/problems/anagram-groups/question
"""
````

````python
from typing import List
from collections import defaultdict

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        anagram_map = defaultdict(list)
        for s in strs:
            # lists are not hashable (cannot be key)
            key = tuple(sorted(s))
            anagram_map[key].append(s)
        return list(anagram_map.values())
````

````python
""" 
Time Complexity: 
O(m⋅nlogn), where m is the number of strings and n is the average length of a string (due to sorting).
Space Complexity: O(m⋅n) to store the dictionary.
"""
````

## Top K Frequent Elements

````python
""" 
Given an integer array nums and an integer k, return the k most frequent elements within the array.

The test cases are generated such that the answer is always unique.

You may return the output in any order.

Example 1:

Input: nums = [1,2,2,3,3,3], k = 2

Output: [2,3]
Example 2:

Input: nums = [7,7], k = 1

Output: [7]
Constraints:

1 <= nums.length <= 10^4.
-1000 <= nums[i] <= 1000
1 <= k <= number of distinct elements in nums.

https://neetcode.io/problems/top-k-elements-in-list/question?list=blind75
"""
````

````python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = {}
        for num in nums:
            count[num] = count.get(num, 0) + 1
        freq = [[] for i in range(len(nums)+1)]
        for n, c in count.items():
            freq[c].append(n)
        res = []
        for i in range(len(freq)-1, 0, -1):
            for n in freq[i]:
                res.append(n)
                if len(res) >= k:
                    return res
        return res
````

````python
"""  
Time Complexity: 
O(N). We iterate through the array to count, then iterate through the buckets.
Space Complexity: 
O(N). To store the hash map and the bucket array.
"""
````

## Encode and Decode Strings

````python
""" 
Design an algorithm to encode a list of strings to a single string. 
The encoded string is then decoded back to the original list of strings.

Please implement encode and decode

Example 1:

Input: ["neet","code","love","you"]

Output:["neet","code","love","you"]
Example 2:

Input: ["we","say",":","yes"]

Output: ["we","say",":","yes"]
Constraints:

0 <= strs.length < 100
0 <= strs[i].length < 200
strs[i] contains only UTF-8 characters.

"""
````

````python
class Solution:

    def encode(self, strs: List[str]) -> str:
        res = ""
        for s in strs:
            res += str(len(s)) + "#" + s
        return res

    def decode(self, s: str) -> List[str]:
        res = []
        i = 0
        while i < len(s):
            j = i
            while s[j] != '#':
                j += 1
            length = int(s[i:j])
            res.append(s[j + 1 : j + 1 + length])
            i = j + 1 + length
        return res
````

````python
""" 

### Walkthrough

**1. Encode `["neet", "code"]`**
*   `"neet"` (len 4) $\to$ becomes `"4#neet"`
*   `"code"` (len 4) $\to$ becomes `"4#code"`
*   **Result:** `"4#neet4#code"`

**2. Decode `"4#neet4#code"`**
*   **First Iteration:**
    *   `i` starts at 0.
    *   Find `#` starting from `i`. It is found at index 1.
    *   Read length between `i` and `#`: `s[0:1]` is `"4"`.
    *   Read 4 characters after the `#`: `"neet"`.
    *   Add `"neet"` to result.
    *   Update `i` to index 6 (start of next segment).
*   **Second Iteration:**
    *   Find `#` starting from `i`. It is found at index 7.
    *   Read length between `i` and `#`: `s[6:7]` is `"4"`.
    *   Read 4 characters after the `#`: `"code"`.
    *   Add `"code"` to result.
    *   Update `i` to index 12.
*   **Loop Ends:** `i` is no longer less than string length.

### Complexity Analysis
*   **Time Complexity:** $O(N)$, where $N$ is the total number of characters across all strings. We iterate through the characters once for encoding and once for decoding.
*   **Space Complexity:** $O(N)$ to hold the encoded string and the decoded list.

"""
````

## Products of Array Except Self

````python
""" 
Given an integer array nums, 
return an array output where output[i] is the product of all the elements of nums except nums[i].

Each product is guaranteed to fit in a 32-bit integer.

Follow-up: Could you solve it in 
O(n) time without using the division operation?

Example 1:

Input: nums = [1,2,4,6]

Output: [48,24,12,8]
Example 2:

Input: nums = [-1,0,1,2,3]

Output: [0,-6,0,0,0]
Constraints:

2 <= nums.length <= 1000
-20 <= nums[i] <= 20

https://neetcode.io/problems/products-of-array-discluding-self/question?list=blind75
"""
````

````python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        res = [1]*n
        prefix = 1
        for i in range(n):
            res[i] = prefix
            prefix *= nums[i]
        postfix = 1
        for i in range(n-1, -1, -1):
            res[i] *= postfix
            postfix *= nums[i]
        return res
````

This is a classic problem that requires calculating the "Prefix Product" (product of all numbers to the left) and the "Suffix Product" (product of all numbers to the right) for every index.

The standard solution uses $O(N)$ time and $O(1)$ extra space (excluding the output array).

### The Logic

For an array `[a, b, c, d]`:
1.  **Pass 1 (Left-to-Right):** Calculate products of all elements to the **left** of the current index.
    *   Result becomes: `[1, a, ab, abc]`
2.  **Pass 2 (Right-to-Left):** Calculate products of all elements to the **right** and multiply them by the existing value in the result array.
    *   Result becomes: `[1*bcd, a*cd, ab*d, abc*1]`

### Python Solution

```python
from typing import List

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        res = [1] * n
        
        # Pass 1: Calculate Prefix Products
        # Store the product of all elements to the left of i
        prefix = 1
        for i in range(n):
            res[i] = prefix
            prefix *= nums[i]
            
        # Pass 2: Calculate Postfix Products
        # Multiply the already stored prefix product by the product of all elements to the right
        postfix = 1
        for i in range(n - 1, -1, -1):
            res[i] *= postfix
            postfix *= nums[i]
            
        return res
```

### Walkthrough (Example 1: `[1, 2, 4, 6]`)

**Pass 1 (Left to Right):**
*   Initialize `res` = `[1, 1, 1, 1]`
*   `i=0`: `res[0]` = 1 (prefix). New prefix: $1 \times 1 = 1$
*   `i=1`: `res[1]` = 1 (prefix). New prefix: $1 \times 2 = 2$
*   `i=2`: `res[2]` = 2 (prefix). New prefix: $2 \times 4 = 8$
*   `i=3`: `res[3]` = 8 (prefix). New prefix: $8 \times 6 = 48$
*   **End of Pass 1:** `res` is `[1, 1, 2, 8]`

**Pass 2 (Right to Left):**
*   `postfix` starts at 1.
*   `i=3`: `res[3]` becomes $8 \times 1 = 8$. New postfix: $1 \times 6 = 6$
*   `i=2`: `res[2]` becomes $2 \times 6 = 12$. New postfix: $6 \times 4 = 24$
*   `i=1`: `res[1]` becomes $1 \times 24 = 24$. New postfix: $24 \times 2 = 48$
*   `i=0`: `res[0]` becomes $1 \times 48 = 48$. New postfix: $48 \times 1 = 48$
*   **Result:** `[48, 24, 12, 8]`

### Complexity Analysis
*   **Time Complexity:** $O(N)$. We iterate through the array exactly twice.
*   **Space Complexity:** $O(1)$ (extra space). The problem statement typically counts the output array as free space. If we counted the output array, it would be $O(N)$, but we use no auxiliary data structures like Hash Maps or temporary arrays.

## Longest Consecutive Sequence


````python
""" 
Given an array of integers nums, 
return the length of the longest consecutive sequence of elements that can be formed.

A consecutive sequence is a sequence of elements in which each element is exactly 1 greater 
than the previous element. The elements do not have to be consecutive in the original array.

You must write an algorithm that runs in O(n) time.

Example 1:

Input: nums = [2,20,4,10,3,4,5]

Output: 4
Explanation: The longest consecutive sequence is [2, 3, 4, 5].

Example 2:

Input: nums = [0,3,2,5,4,6,1,1]

Output: 7
Constraints:

0 <= nums.length <= 1000
-10^9 <= nums[i] <= 10^9
"""
````

````python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        num_set = set(nums)
        longest = 0
        for num in num_set:
            if (num-1) not in num_set:
                length = 1
                while (num + length) in num_set:
                    length += 1
                longest = max(longest, length)
        return longest 
````

The most efficient way to solve this problem is using a **Hash Set**.

### The Logic ($O(n)$ Approach)

1.  **Eliminate Duplicates & Enable Fast Lookups:** Convert the array into a Set. This allows us to check if a number exists in $O(1)$ time and removes duplicates.
2.  **Identify Sequence Starts:** We iterate through the set. We only attempt to build a sequence if the current number is the **start** of a sequence.
    *   How do we know if a number `x` is the start? If `x - 1` is **not** in the set.
    *   If `x - 1` *is* in the set, we skip `x` because it is already part of a sequence we count starting from somewhere else.
3.  **Count the Sequence:** Once we find a start, we loop upwards (`x+1`, `x+2`, etc.) checking the set until the sequence breaks.

### Solution

```python
from typing import List

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        numSet = set(nums)
        longest = 0

        for n in numSet:
            # check if its the start of a sequence
            if (n - 1) not in numSet:
                length = 1
                while (n + length) in numSet:
                    length += 1
                longest = max(length, longest)
                
        return longest
```

### Walkthrough (Example: `[2, 20, 4, 10, 3, 4, 5]`)

1.  **Create Set:** `{2, 20, 4, 10, 3, 5}` (duplicates removed).
2.  **Iterate:**
    *   **Pick 2:** Is `1` in the set? **No**. `2` is a start.
        *   Count: 2 exists, 3 exists, 4 exists, 5 exists, 6 does not.
        *   Length is 4. `longest = 4`.
    *   **Pick 20:** Is `19` in the set? **No**. `20` is a start.
        *   Count: 20 exists, 21 does not.
        *   Length is 1. `longest` remains 4.
    *   **Pick 4:** Is `3` in the set? **Yes**. Skip (because we already counted 4 when we started at 2).
    *   **Pick 10:** Is `9` in the set? **No**. `10` is a start.
        *   Length is 1. `longest` remains 4.
    *   **Pick 3:** Is `2` in the set? **Yes**. Skip.
    *   **Pick 5:** Is `4` in the set? **Yes**. Skip.

**Result:** 4

### Complexity Analysis
*   **Time Complexity:** $O(n)$.
    *   Converting to a set takes $O(n)$.
    *   The loop iterates $n$ times.
    *   Crucially, the inner `while` loop runs only when `n` is the *start* of a sequence. This means each number is visited inside the `while` loop exactly once across the entire execution. Thus, the total operations are linear.
*   **Space Complexity:** $O(n)$ to store the set.

````python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        if not nums:
            return 0

        num_set = sorted(set(nums))
        longest = 1
        current_streak = 1

        for i in range(1, len(num_set)):
            if num_set[i-1] + 1 == num_set[i]:
                current_streak += 1
            else:
                longest = max(longest, current_streak)
                current_streak = 1

        return max(current_streak, longest) 
````


题目明确要求算法的时间复杂度必须是 **O(n)**。

如果你对 `numSet`（或者直接对原数组）进行排序，排序这个操作本身的最低时间复杂度是 **O(n log n)**。因此，虽然代码能跑通并通过测试用例（如果测试数据量不够大），但在严格的算法面试或判题标准中，这属于**不合格**的解法，因为它比 O(n) 慢。

不过，在面试中，如果你一时想不出 O(n) 的解法，**先写出排序解法（O(n log n)）通常是比写不出来要好的**。

### 排序解法代码 (O(n log n))

如果你不在意 O(n) 的限制，只想解出答案，代码会非常直观：

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        if not nums:
            return 0
            
        # 1. 去重并排序
        # 排序的时间复杂度是 O(n log n)
        nums = sorted(set(nums))
        
        longest = 1
        current_streak = 1
        
        # 2. 线性扫描
        for i in range(1, len(nums)):
            # 如果当前数字等于前一个数字 + 1
            if nums[i] == nums[i-1] + 1:
                current_streak += 1
            else:
                # 序列断了，更新最大值，重置当前计数
                longest = max(longest, current_streak)
                current_streak = 1
                
        return max(longest, current_streak)
```

### 为什么题目要求 O(n)？

这道题的考点正是**如何利用空间换时间**。

*   **排序法 (Sorting):** 
    *   **时间:** $O(n \log n)$ —— 瓶颈在排序。
    *   **空间:** $O(1)$ 或 $O(n)$ (取决于排序算法的具体实现)。
*   **哈希集合法 (Hash Set):** 
    *   **时间:** $O(n)$ —— 只需要遍历数组几次（虽然有两层循环，但因为有 `if (n-1) not in numSet` 的判断，每个元素实际上只会被访问常数次）。
    *   **空间:** $O(n)$ —— 需要额外的 Set 来存储数据。

**总结：**
如果面试官问“能不能排序”，你可以回答：“逻辑上可以解决问题，但排序会导致时间复杂度变成 O(n log n)，不满足题目 O(n) 的要求。所以为了达到 O(n)，我们需要使用哈希表（Set）。”
