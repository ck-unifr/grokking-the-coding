# Array and Hashing

## Contains Duplicate



````python
"""  
Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:

Input: nums = [1, 2, 3, 3]

Output: true

Example 2:

Input: nums = [1, 2, 3, 4]

Output: false

https://neetcode.io/problems/duplicate-integer/question?list=blind75
"""
````

````python
from typing import List 


class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()
        for num in nums:
            if num in seen:
                return True
            seen.add(num)
        return False
````

````python
"""  
Time Complexity: 
O(n)
 - We iterate through the array once.
Space Complexity: 
O(n)
 - We might store up to n elements in the set.
"""
````

## Valid Anagram



````python
"""  
Given two strings s and t, return true if the two strings are anagrams of each other, 
otherwise return false.

An anagram is a string that contains the exact same characters as another string, 
but the order of the characters can be different.

Example 1:

Input: s = "racecar", t = "carrace"

Output: true
Example 2:

Input: s = "jar", t = "jam"

Output: false
Constraints:

s and t consist of lowercase English letters.
"""
````

````python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        return sorted(s) == sorted(t)
    
````

````python
"""  
Time Complexity: 
O(N)
 (we iterate through the strings once).
Space Complexity: 
O(1)
 (since there are only 26 lowercase English letters, the dictionary size has a constant limit).
"""
````

## Group Anagrams

````python
"""  
Given an array of strings strs, group all anagrams together into sublists. 
You may return the output in any order.

An anagram is a string that contains the exact same characters as another string, 
but the order of the characters can be different.

Example 1:

Input: strs = ["act","pots","tops","cat","stop","hat"]

Output: [["hat"],["act", "cat"],["stop", "pots", "tops"]]
Example 2:

Input: strs = ["x"]

Output: [["x"]]
Example 3:

Input: strs = [""]

Output: [[""]]
Constraints:

1 <= strs.length <= 1000.
0 <= strs[i].length <= 100
strs[i] is made up of lowercase English letters.


https://neetcode.io/problems/anagram-groups/question
"""
````

````python
from typing import List
from collections import defaultdict

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        anagram_map = defaultdict(list)
        for s in strs:
            # lists are not hashable (cannot be key)
            key = tuple(sorted(s))
            anagram_map[key].append(s)
        return list(anagram_map.values())
````

````python
""" 
Time Complexity: 
O(m⋅nlogn), where m is the number of strings and n is the average length of a string (due to sorting).
Space Complexity: O(m⋅n) to store the dictionary.
"""
````

## Top K Frequent Elements

````python
""" 
Given an integer array nums and an integer k, return the k most frequent elements within the array.

The test cases are generated such that the answer is always unique.

You may return the output in any order.

Example 1:

Input: nums = [1,2,2,3,3,3], k = 2

Output: [2,3]
Example 2:

Input: nums = [7,7], k = 1

Output: [7]
Constraints:

1 <= nums.length <= 10^4.
-1000 <= nums[i] <= 1000
1 <= k <= number of distinct elements in nums.

https://neetcode.io/problems/top-k-elements-in-list/question?list=blind75
"""
````

````python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = {}
        for num in nums:
            count[num] = count.get(num, 0) + 1
        freq = [[] for i in range(len(nums)+1)]
        for n, c in count.items():
            freq[c].append(n)
        res = []
        for i in range(len(freq)-1, 0, -1):
            for n in freq[i]:
                res.append(n)
                if len(res) >= k:
                    return res
        return res
````

````python
"""  
Time Complexity: 
O(N). We iterate through the array to count, then iterate through the buckets.
Space Complexity: 
O(N). To store the hash map and the bucket array.
"""
````

## Encode and Decode Strings

````python
""" 
Design an algorithm to encode a list of strings to a single string. 
The encoded string is then decoded back to the original list of strings.

Please implement encode and decode

Example 1:

Input: ["neet","code","love","you"]

Output:["neet","code","love","you"]
Example 2:

Input: ["we","say",":","yes"]

Output: ["we","say",":","yes"]
Constraints:

0 <= strs.length < 100
0 <= strs[i].length < 200
strs[i] contains only UTF-8 characters.

"""
````

````python
class Solution:

    def encode(self, strs: List[str]) -> str:
        res = ""
        for s in strs:
            res += str(len(s)) + "#" + s
        return res

    def decode(self, s: str) -> List[str]:
        res = []
        i = 0
        while i < len(s):
            j = i
            while s[j] != '#':
                j += 1
            length = int(s[i:j])
            res.append(s[j + 1 : j + 1 + length])
            i = j + 1 + length
        return res
````

````python
""" 

### Walkthrough

**1. Encode `["neet", "code"]`**
*   `"neet"` (len 4) $\to$ becomes `"4#neet"`
*   `"code"` (len 4) $\to$ becomes `"4#code"`
*   **Result:** `"4#neet4#code"`

**2. Decode `"4#neet4#code"`**
*   **First Iteration:**
    *   `i` starts at 0.
    *   Find `#` starting from `i`. It is found at index 1.
    *   Read length between `i` and `#`: `s[0:1]` is `"4"`.
    *   Read 4 characters after the `#`: `"neet"`.
    *   Add `"neet"` to result.
    *   Update `i` to index 6 (start of next segment).
*   **Second Iteration:**
    *   Find `#` starting from `i`. It is found at index 7.
    *   Read length between `i` and `#`: `s[6:7]` is `"4"`.
    *   Read 4 characters after the `#`: `"code"`.
    *   Add `"code"` to result.
    *   Update `i` to index 12.
*   **Loop Ends:** `i` is no longer less than string length.

### Complexity Analysis
*   **Time Complexity:** $O(N)$, where $N$ is the total number of characters across all strings. We iterate through the characters once for encoding and once for decoding.
*   **Space Complexity:** $O(N)$ to hold the encoded string and the decoded list.

"""
````

## Products of Array Except Self

````python
""" 
Given an integer array nums, 
return an array output where output[i] is the product of all the elements of nums except nums[i].

Each product is guaranteed to fit in a 32-bit integer.

Follow-up: Could you solve it in 
O(n) time without using the division operation?

Example 1:

Input: nums = [1,2,4,6]

Output: [48,24,12,8]
Example 2:

Input: nums = [-1,0,1,2,3]

Output: [0,-6,0,0,0]
Constraints:

2 <= nums.length <= 1000
-20 <= nums[i] <= 20

https://neetcode.io/problems/products-of-array-discluding-self/question?list=blind75
"""
````

````python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        res = [1]*n
        prefix = 1
        for i in range(n):
            res[i] = prefix
            prefix *= nums[i]
        postfix = 1
        for i in range(n-1, -1, -1):
            res[i] *= postfix
            postfix *= nums[i]
        return res
````

This is a classic problem that requires calculating the "Prefix Product" (product of all numbers to the left) and the "Suffix Product" (product of all numbers to the right) for every index.

The standard solution uses $O(N)$ time and $O(1)$ extra space (excluding the output array).

### The Logic

For an array `[a, b, c, d]`:
1.  **Pass 1 (Left-to-Right):** Calculate products of all elements to the **left** of the current index.
    *   Result becomes: `[1, a, ab, abc]`
2.  **Pass 2 (Right-to-Left):** Calculate products of all elements to the **right** and multiply them by the existing value in the result array.
    *   Result becomes: `[1*bcd, a*cd, ab*d, abc*1]`

### Python Solution

```python
from typing import List

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        res = [1] * n
        
        # Pass 1: Calculate Prefix Products
        # Store the product of all elements to the left of i
        prefix = 1
        for i in range(n):
            res[i] = prefix
            prefix *= nums[i]
            
        # Pass 2: Calculate Postfix Products
        # Multiply the already stored prefix product by the product of all elements to the right
        postfix = 1
        for i in range(n - 1, -1, -1):
            res[i] *= postfix
            postfix *= nums[i]
            
        return res
```

### Walkthrough (Example 1: `[1, 2, 4, 6]`)

**Pass 1 (Left to Right):**
*   Initialize `res` = `[1, 1, 1, 1]`
*   `i=0`: `res[0]` = 1 (prefix). New prefix: $1 \times 1 = 1$
*   `i=1`: `res[1]` = 1 (prefix). New prefix: $1 \times 2 = 2$
*   `i=2`: `res[2]` = 2 (prefix). New prefix: $2 \times 4 = 8$
*   `i=3`: `res[3]` = 8 (prefix). New prefix: $8 \times 6 = 48$
*   **End of Pass 1:** `res` is `[1, 1, 2, 8]`

**Pass 2 (Right to Left):**
*   `postfix` starts at 1.
*   `i=3`: `res[3]` becomes $8 \times 1 = 8$. New postfix: $1 \times 6 = 6$
*   `i=2`: `res[2]` becomes $2 \times 6 = 12$. New postfix: $6 \times 4 = 24$
*   `i=1`: `res[1]` becomes $1 \times 24 = 24$. New postfix: $24 \times 2 = 48$
*   `i=0`: `res[0]` becomes $1 \times 48 = 48$. New postfix: $48 \times 1 = 48$
*   **Result:** `[48, 24, 12, 8]`

### Complexity Analysis
*   **Time Complexity:** $O(N)$. We iterate through the array exactly twice.
*   **Space Complexity:** $O(1)$ (extra space). The problem statement typically counts the output array as free space. If we counted the output array, it would be $O(N)$, but we use no auxiliary data structures like Hash Maps or temporary arrays.

## Longest Consecutive Sequence


````python
""" 
Given an array of integers nums, 
return the length of the longest consecutive sequence of elements that can be formed.

A consecutive sequence is a sequence of elements in which each element is exactly 1 greater 
than the previous element. The elements do not have to be consecutive in the original array.

You must write an algorithm that runs in O(n) time.

Example 1:

Input: nums = [2,20,4,10,3,4,5]

Output: 4
Explanation: The longest consecutive sequence is [2, 3, 4, 5].

Example 2:

Input: nums = [0,3,2,5,4,6,1,1]

Output: 7
Constraints:

0 <= nums.length <= 1000
-10^9 <= nums[i] <= 10^9
"""
````

````python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        num_set = set(nums)
        longest = 0
        for num in num_set:
            if (num-1) not in num_set:
                length = 1
                while (num + length) in num_set:
                    length += 1
                longest = max(longest, length)
        return longest 
````

The most efficient way to solve this problem is using a **Hash Set**.

### The Logic ($O(n)$ Approach)

1.  **Eliminate Duplicates & Enable Fast Lookups:** Convert the array into a Set. This allows us to check if a number exists in $O(1)$ time and removes duplicates.
2.  **Identify Sequence Starts:** We iterate through the set. We only attempt to build a sequence if the current number is the **start** of a sequence.
    *   How do we know if a number `x` is the start? If `x - 1` is **not** in the set.
    *   If `x - 1` *is* in the set, we skip `x` because it is already part of a sequence we count starting from somewhere else.
3.  **Count the Sequence:** Once we find a start, we loop upwards (`x+1`, `x+2`, etc.) checking the set until the sequence breaks.

### Solution

```python
from typing import List

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        numSet = set(nums)
        longest = 0

        for n in numSet:
            # check if its the start of a sequence
            if (n - 1) not in numSet:
                length = 1
                while (n + length) in numSet:
                    length += 1
                longest = max(length, longest)
                
        return longest
```

### Walkthrough (Example: `[2, 20, 4, 10, 3, 4, 5]`)

1.  **Create Set:** `{2, 20, 4, 10, 3, 5}` (duplicates removed).
2.  **Iterate:**
    *   **Pick 2:** Is `1` in the set? **No**. `2` is a start.
        *   Count: 2 exists, 3 exists, 4 exists, 5 exists, 6 does not.
        *   Length is 4. `longest = 4`.
    *   **Pick 20:** Is `19` in the set? **No**. `20` is a start.
        *   Count: 20 exists, 21 does not.
        *   Length is 1. `longest` remains 4.
    *   **Pick 4:** Is `3` in the set? **Yes**. Skip (because we already counted 4 when we started at 2).
    *   **Pick 10:** Is `9` in the set? **No**. `10` is a start.
        *   Length is 1. `longest` remains 4.
    *   **Pick 3:** Is `2` in the set? **Yes**. Skip.
    *   **Pick 5:** Is `4` in the set? **Yes**. Skip.

**Result:** 4

### Complexity Analysis
*   **Time Complexity:** $O(n)$.
    *   Converting to a set takes $O(n)$.
    *   The loop iterates $n$ times.
    *   Crucially, the inner `while` loop runs only when `n` is the *start* of a sequence. This means each number is visited inside the `while` loop exactly once across the entire execution. Thus, the total operations are linear.
*   **Space Complexity:** $O(n)$ to store the set.

````python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        if not nums:
            return 0

        num_set = sorted(set(nums))
        longest = 1
        current_streak = 1

        for i in range(1, len(num_set)):
            if num_set[i-1] + 1 == num_set[i]:
                current_streak += 1
            else:
                longest = max(longest, current_streak)
                current_streak = 1

        return max(current_streak, longest) 
````


题目明确要求算法的时间复杂度必须是 **O(n)**。

如果你对 `numSet`（或者直接对原数组）进行排序，排序这个操作本身的最低时间复杂度是 **O(n log n)**。因此，虽然代码能跑通并通过测试用例（如果测试数据量不够大），但在严格的算法面试或判题标准中，这属于**不合格**的解法，因为它比 O(n) 慢。

不过，在面试中，如果你一时想不出 O(n) 的解法，**先写出排序解法（O(n log n)）通常是比写不出来要好的**。

### 排序解法代码 (O(n log n))

如果你不在意 O(n) 的限制，只想解出答案，代码会非常直观：

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        if not nums:
            return 0
            
        # 1. 去重并排序
        # 排序的时间复杂度是 O(n log n)
        nums = sorted(set(nums))
        
        longest = 1
        current_streak = 1
        
        # 2. 线性扫描
        for i in range(1, len(nums)):
            # 如果当前数字等于前一个数字 + 1
            if nums[i] == nums[i-1] + 1:
                current_streak += 1
            else:
                # 序列断了，更新最大值，重置当前计数
                longest = max(longest, current_streak)
                current_streak = 1
                
        return max(longest, current_streak)
```

### 为什么题目要求 O(n)？

这道题的考点正是**如何利用空间换时间**。

*   **排序法 (Sorting):** 
    *   **时间:** $O(n \log n)$ —— 瓶颈在排序。
    *   **空间:** $O(1)$ 或 $O(n)$ (取决于排序算法的具体实现)。
*   **哈希集合法 (Hash Set):** 
    *   **时间:** $O(n)$ —— 只需要遍历数组几次（虽然有两层循环，但因为有 `if (n-1) not in numSet` 的判断，每个元素实际上只会被访问常数次）。
    *   **空间:** $O(n)$ —— 需要额外的 Set 来存储数据。

**总结：**
如果面试官问“能不能排序”，你可以回答：“逻辑上可以解决问题，但排序会导致时间复杂度变成 O(n log n)，不满足题目 O(n) 的要求。所以为了达到 O(n)，我们需要使用哈希表（Set）。”

## Valid Palindrome

````python
"""  
Given a string s, return true if it is a palindrome, otherwise return false.

A palindrome is a string that reads the same forward and backward. 
It is also case-insensitive and ignores all non-alphanumeric characters.

Note: Alphanumeric characters consist of letters (A-Z, a-z) and numbers (0-9).

Example 1:

Input: s = "Was it a car or a cat I saw?"

Output: true
Explanation: After considering only alphanumerical characters we have "wasitacaroracatisaw", which is a palindrome.

Example 2:

Input: s = "tab a cat"

Output: false
Explanation: "tabacat" is not a palindrome.

Constraints:

1 <= s.length <= 1000
s is made up of only printable ASCII characters.

https://neetcode.io/problems/is-palindrome/question?list=blind75

"""
````

````python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        l, r = 0, len(s)-1
        while l < r:
            if not s[l].isalnum():
                l += 1
                continue
            if not s[r].isalnum():
                r -= 1
                continue
            if s[l].lower() != s[r].lower():
                return False
            l += 1
            r -= 1
        return True
````

Here are the two best ways to solve this problem.

### Approach 1: Two Pointers (Space Optimized)
This is the standard algorithm for interviews because it solves the problem in-place with **$O(1)$ extra memory**.

We use two pointers: `left` starting at the beginning and `right` starting at the end. We move them toward the center, skipping non-alphanumeric characters.

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        l, r = 0, len(s) - 1

        while l < r:
            # Move left pointer forward until we find an alphanumeric char
            if not s[l].isalnum():
                l += 1
                continue
            
            # Move right pointer backward until we find an alphanumeric char
            if not s[r].isalnum():
                r -= 1
                continue
            
            # Compare characters (case-insensitive)
            if s[l].lower() != s[r].lower():
                return False
            
            # After a successful match, move both pointers inward
            l += 1
            r -= 1
            
        return True
```
*   **Time Complexity:** $O(n)$
*   **Space Complexity:** $O(1)$

---

### Approach 2: Filter and Reverse (Pythonic)
This approach cleans the string first by removing non-alphanumeric characters and converting to lowercase, then compares the new string to its reverse. It is shorter to write but uses extra memory.

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        # Create a new string with only alphanumeric chars, converted to lowercase
        new_s = ''.join(char.lower() for char in s if char.isalnum())
        
        # Check if the string equals its reverse
        return new_s == new_s[::-1]
```
*   **Time Complexity:** $O(n)$
*   **Space Complexity:** $O(n)$ (creates a new string)

### Explanation of Helper Methods used:
*   **`isalnum()`**: Returns `True` if the character is a letter (a-z) or a number (0-9). It ignores spaces and punctuation.
*   **`lower()`**: Converts the character to lowercase so 'A' equals 'a'.
*   **`[::-1]`**: Python's slicing syntax to create a reversed copy of a string.

## 3Sum

````python
"""  
https://neetcode.io/problems/three-integer-sum/question?list=blind75

Given an integer array nums, 
return all the triplets [nums[i], nums[j], nums[k]] where nums[i] + nums[j] + nums[k] == 0, 
and the indices i, j and k are all distinct.

The output should not contain any duplicate triplets. 
You may return the output and the triplets in any order.

Example 1:

Input: nums = [-1,0,1,2,-1,-4]

Output: [[-1,-1,2],[-1,0,1]]
Explanation:
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].

Example 2:

Input: nums = [0,1,1]

Output: []
Explanation: The only possible triplet does not sum up to 0.

Example 3:

Input: nums = [0,0,0]

Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.

Constraints:

3 <= nums.length <= 1000
-10^5 <= nums[i] <= 10^5

"""
````

````python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        res = []
        
        for i in range(len(nums) - 2):
            # 剪枝：如果当前数字大于0，后面不可能凑出0
            if nums[i] > 0: break 
            # 去重：跳过重复的起始数字
            if i > 0 and nums[i] == nums[i-1]: continue
            l, r = i + 1, len(nums) - 1
            while l < r:
                s = nums[i] + nums[l] + nums[r]
                if s == 0:
                    res.append([nums[i], nums[l], nums[r]])
                    # 核心去重逻辑：在一行内完成指针移动
                    while l < r and nums[l] == nums[l+1]: l += 1
                    while l < r and nums[r] == nums[r-1]: r -= 1
                    l += 1; r -= 1
                elif s < 0:
                    l += 1
                else:
                    r -= 1
        return res
        
````

当然可以。如果追求代码的**简洁性**（减少缩进和冗余逻辑），同时保持 **$O(N^2)$** 的效率，有以下两种优化写法。

### 写法 1：利用 Set 自动去重（代码最少）

这种写法利用 Python 的 `set` 来处理结果的去重，从而省去了手动编写 `while` 循环跳过重复元素的代码。逻辑非常清晰。

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        res = set()  # 使用集合自动去重
        
        for i in range(len(nums) - 2):
            l, r = i + 1, len(nums) - 1
            while l < r:
                s = nums[i] + nums[l] + nums[r]
                if s == 0:
                    res.add((nums[i], nums[l], nums[r])) # 元组可哈希，存入set
                    l += 1
                    r -= 1
                elif s < 0:
                    l += 1
                else:
                    r -= 1
                    
        return [list(x) for x in res]
```

*   **优点**：代码极短，不需要处理复杂的边界和重复跳过逻辑。
*   **缺点**：虽然通过了测试，但因为没有在发现答案时立即跳过连续相同的数字，会多做一些无用的计算，且 Set 会稍微多占用一点内存。

---

### 写法 2：标准双指针精简版（面试推荐）

这是标准解法的“紧凑版”。保留了手动去重的逻辑（效率最高），但把代码压缩得更紧凑。这是**面试中最完美**的平衡：既简洁，又展示了对性能的极致追求。

```python
                if s == 0:
                    res.append([nums[i], nums[l], nums[r]])
                    # 核心去重逻辑：在一行内完成指针移动
                    while l < r and nums[l] == nums[l+1]: l += 1
                    while l < r and nums[r] == nums[r-1]: r -= 1
                    l += 1; r -= 1
                elif s < 0:
                    l += 1
                else:
                    r -= 1
        return res
```

### 为什么不能用一行代码 (One-Liner)？
理论上可以使用 `itertools.combinations` 写成一行：
```python
# 千万别在面试写这个，会超时 (Time Limit Exceeded)

```

## Container With Most Water

````python
""" 
https://neetcode.io/problems/max-water-container/question?list=blind75

You are given an integer array heights where heights[i] represents the height of the 
ith bar.

You may choose any two bars to form a container. Return the maximum amount of water a container can store.
"""
````

````python
class Solution:
    def maxArea(self, heights: List[int]) -> int:
        l, r = 0, len(heights) - 1
        res = 0
        while l < r:
            res = max(res, min(heights[l], heights[r])*(r - l))
            if heights[l] < heights[r]:
                l += 1
            else:
                r -= 1
        return res
````

This problem is known as **Container With Most Water**.

The most efficient solution uses the **Two Pointers** technique.

### The Logic ($O(N)$)

1.  **Start Max Width:** Place one pointer at the beginning (`l`) and one at the end (`r`). This gives us the maximum possible width.
2.  **Calculate Area:** Calculate the area formed by these two lines: `min(height[l], height[r]) * (r - l)`. Update the maximum result.
3.  **Greedy Move:** To potentially find a larger area, we need to find a taller line to compensate for the reduction in width as we move pointers inward.
    *   **Move the shorter line.** Why? Because the area is limited by the shorter line. If we move the taller line, the height remains limited by the shorter line (or becomes even smaller), but the width decreases, guaranteeing a smaller area.
    *   If `height[l] < height[r]`, move `l` to the right.
    *   Otherwise, move `r` to the left.
4.  **Repeat:** Continue until the pointers meet.

### Solution in Python

```python
from typing import List

class Solution:
    def maxArea(self, heights: List[int]) -> int:
        l, r = 0, len(heights) - 1
        res = 0
        
        while l < r:
            # Calculate height of the container (limited by the shorter side)
            area_height = min(heights[l], heights[r])
            # Calculate width
            area_width = r - l
            
            # Update max area
            res = max(res, area_height * area_width)
            
            # Move the pointer of the shorter line inward
            if heights[l] < heights[r]:
                l += 1
            else:
                r -= 1
                
        return res
```

### Walkthrough (Example: `[1, 8, 6, 2, 5, 4, 8, 3, 7]`)

1.  **l=0 (1), r=8 (7):**
    *   Width = 8. Height = min(1, 7) = 1.
    *   Area = 8. `res` = 8.
    *   `height[l]` is smaller. Move `l` $\to$ 1.
2.  **l=1 (8), r=8 (7):**
    *   Width = 7. Height = min(8, 7) = 7.
    *   Area = 49. `res` = 49.
    *   `height[r]` is smaller. Move `r` $\to$ 7.
3.  **l=1 (8), r=7 (3):**
    *   Width = 6. Height = 3.
    *   Area = 18. `res` remains 49.
    *   `height[r]` is smaller. Move `r` $\to$ 6.
4.  **...Continue until pointers meet.**

### Complexity Analysis
*   **Time Complexity:** $O(N)$. We visit every element at most once.
*   **Space Complexity:** $O(1)$. We only use variables for pointers and the result.

## Best Time to Buy and Sell Stock


````python
""" 
https://neetcode.io/problems/buy-and-sell-crypto/question

You are given an integer array prices where prices[i] is the price of NeetCoin on the ith day.

You may choose a single day to buy one NeetCoin and choose a different day in the future to sell it.

Return the maximum profit you can achieve. You may choose to not make any transactions, 
in which case the profit would be 0.

Example 1:

Input: prices = [10,1,5,6,7,1]

Output: 6
Explanation: Buy prices[1] and sell prices[4], profit = 7 - 1 = 6.

Example 2:

Input: prices = [10,8,7,5,2]

Output: 0
Explanation: No profitable transactions can be made, thus the max profit is 0.

Constraints:

1 <= prices.length <= 100
0 <= prices[i] <= 100

"""
````

````python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        max_profit = 0
        min_price = float('inf')
        for price in prices:
            min_price = min(price, min_price)
            max_profit = max(max_profit, (price - min_price))
        return max_profit
````

我们可以去掉显式的双指针（索引操作），直接在遍历过程中维护一个**“最低买入价”**（`min_price`）。

这种写法逻辑更直观：对于每一天的价格，我们只关心两件事：
1.  现在的价格是不是比我之前见过的都低？如果是，更新最低价。
2.  如果我今天卖出（当前价格 - 最低价），能不能赚更多？如果是，更新最大利润。

### 写法 1：贪心迭代（推荐，最优雅）

这种写法非常 Pythonic，代码行数少，且完全保留了 $O(N)$ 的时间效率和 $O(1)$ 的空间效率。

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_price = float('inf')  # 初始化为一个极大值
        max_profit = 0
        
        for p in prices:
            # 1. 更新至今为止的最低价格
            min_price = min(min_price, p)
            
            # 2. 计算如果今天卖出的利润，并更新最大利润
            max_profit = max(max_profit, p - min_price)
            
        return max_profit
```

---

### 写法 2：一行代码 (One-Liner)

如果你想在面试中展示对 Python 标准库 `itertools` 的熟悉程度，或者只是单纯追求代码简短，可以使用 `accumulate`。

`accumulate(prices, min)` 会生成一个序列，表示“截止到当前位置的最小值”。我们将原数组和这个最小值序列 `zip` 起来相减，就能得到每一天卖出的最大潜在利润。

```python
from itertools import accumulate

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        # 逻辑：max(当前价格 - 截止当前的最低价格)
        # prices:             [10, 1, 5, 6, 7, 1]
        # accumulate(..,min): [10, 1, 1, 1, 1, 1]
        # 差值:                [ 0, 0, 4, 5, 6, 0] -> max is 6
        return max(p - m for p, m in zip(prices, accumulate(prices, min))) if prices else 0
```

**总结：**
*   **写法 1** 是最通用的工程写法，清晰易读。
*   **写法 2** 是“炫技”写法，虽然行数最少，但引入了额外的模块，且在某些语言实现中可能会因为生成中间迭代器而略微增加内存开销（尽管在 Python 中 zip/accumulate 是惰性的，空间复杂度依然优秀）。

## Longest Substring Without Repeating Characters

````python
""" 
https://neetcode.io/problems/longest-substring-without-duplicates/question?list=blind75

Given a string s, find the length of the longest substring without duplicate characters.

A substring is a contiguous sequence of characters within a string.

Example 1:

Input: s = "zxyzxyz"

Output: 3
Explanation: The string "xyz" is the longest without duplicate characters.

Example 2:

Input: s = "xxxx"

Output: 1
Constraints:

0 <= s.length <= 1000
s may consist of printable ASCII characters.
"""
````

````python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        char_set = set()
        l = 0
        res = 0
        for r in range(len(s)):
            while s[r] in char_set:
                char_set.remove(s[l])
                l += 1
            char_set.add(s[r])
            res = max(res, r - l + 1)
        return res
````

The most efficient and standard way to solve this problem is using the **Sliding Window** technique combined with a **Hash Set**.

### The Logic ($O(N)$)

We maintain a window defined by two pointers, `l` (left) and `r` (right).
1.  **Expand (`r`):** We iterate through the string with the `r` pointer, adding characters to a `set`.
2.  **Shrink (`l`):** If we encounter a character that is **already in the set**, it means we have a duplicate. We must shrink the window from the left by removing characters and moving `l` forward until the duplicate is removed.
3.  **Track Max:** At every step, update the maximum length found so far (`r - l + 1`).

### Solution in Python

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        charSet = set()
        l = 0
        res = 0
        
        for r in range(len(s)):
            # If the character is already in the set, shrink the window from the left
            while s[r] in charSet:
                charSet.remove(s[l])
                l += 1
            
            # Add the current character to the set
            charSet.add(s[r])
            
            # Update the maximum length found
            res = max(res, r - l + 1)
            
        return res
```

### Walkthrough (Example: `s = "zxyzxyz"`)

1.  **`r=0` ('z')**: Set: `{'z'}`, Max: 1
2.  **`r=1` ('x')**: Set: `{'z', 'x'}`, Max: 2
3.  **`r=2` ('y')**: Set: `{'z', 'x', 'y'}`, Max: 3
4.  **`r=3` ('z')**:
    *   Duplicate found ('z').
    *   **While loop**: Remove `s[l]` ('z'). Move `l` to 1.
    *   Now 'z' is gone from set. Add new 'z'. Set: `{'x', 'y', 'z'}`. Max: 3.
5.  **`r=4` ('x')**:
    *   Duplicate found ('x').
    *   **While loop**: Remove `s[l]` ('x'). Move `l` to 2.
    *   Add new 'x'. Set: `{'y', 'z', 'x'}`. Max: 3.

### Complexity Analysis
*   **Time Complexity:** $O(N)$. In the worst case, each character is visited twice (once by `r` and once by `l`).
*   **Space Complexity:** $O(min(N, M))$, where $M$ is the size of the character set (e.g., 26 for a-z, 128 for ASCII). The set stores unique characters in the current window.

## Longest Repeating Character Replacement

````python
""" 
https://neetcode.io/problems/longest-repeating-substring-with-replacement/question?list=blind75

You are given a string s consisting of only uppercase english characters and an integer k. 
You can choose up to k characters of the string and replace them with any other uppercase English character.

After performing at most k replacements, 
return the length of the longest substring which contains only one distinct character.

Example 1:

Input: s = "XYYX", k = 2

Output: 4
Explanation: Either replace the 'X's with 'Y's, or replace the 'Y's with 'X's.

Example 2:

Input: s = "AAABABB", k = 1

Output: 5
Constraints:

1 <= s.length <= 1000
0 <= k <= s.length
"""
````

````python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        count = {}
        l = 0
        max_fre = 0
        res = 0
        for r in range(len(s)):
            count[s[r]] = count.get(s[r], 0) + 1
            max_fre = max(max_fre, count[s[r]])
            while (r - l + 1 - max_fre) > k:
                count[s[l]] -= 1
                l += 1
            res = max(res, r - l + 1)
        return res
````

This is a classic **Sliding Window** problem.

### The Logic

We want to find the longest window (`substring`) where we can replace at most `k` characters to make all characters in that window the same.

1.  **The Condition:** For any window, let the length be `L` and the count of the **most frequent character** in that window be `max_f`.
    *   The number of characters we need to replace is `L - max_f`.
    *   The window is valid if `L - max_f <= k`.
2.  **Sliding Window:**
    *   We expand the right pointer (`r`) to grow the window.
    *   We track the frequency of characters in the current window.
    *   If the window becomes invalid (i.e., we need more than `k` replacements), we move the left pointer (`l`) to shrink the window until it becomes valid again.

### Solution in Python

```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        count = {}
        res = 0
        l = 0
        max_f = 0
        
        for r in range(len(s)):
            # Add current character to count
            count[s[r]] = count.get(s[r], 0) + 1
            
            # Update the count of the most frequent character in the current window
            max_f = max(max_f, count[s[r]])
            
            # Check if the current window is valid
            # Window Length = (r - l + 1)
            # Replacements needed = Window Length - Most Frequent Char Count
            while (r - l + 1) - max_f > k:
                # The window is invalid, shrink it from the left
                count[s[l]] -= 1
                l += 1
            
            # Update the max length found so far
            res = max(res, r - l + 1)
            
        return res
```

### Walkthrough (Example: `s = "AABABBA", k = 1`)

*   **r=0 ('A')**: `max_f` (A)=1. Len=1. Valid (`1-1 <= 1`). `res`=1.
*   **r=1 ('A')**: `max_f` (A)=2. Len=2. Valid (`2-2 <= 1`). `res`=2.
*   **r=2 ('B')**: `max_f` (A)=2. Len=3. Valid (`3-2 <= 1`). `res`=3.
    *   *Note: We can replace the 1 'B' to get "AAA".*
*   **r=3 ('A')**: `max_f` (A)=3. Len=4. Valid (`4-3 <= 1`). `res`=4.
*   **r=4 ('B')**: `max_f` (A)=3. Len=5. **Invalid** (`5-3 > 1`).
    *   We need 2 replacements (for the two 'B's), but k=1.
    *   **Shrink:** Remove `s[l]` ('A'). `l` becomes 1.
    *   Now window is "ABAB". `max_f` is still technically tracked as high, but conceptually the window is now valid length 4.
*   **r=...**: Continues expanding and shifting.

### Complexity Analysis
*   **Time Complexity:** $O(N)$. The right pointer `r` iterates through the string once. The left pointer `l` also moves at most $N$ times.
*   **Space Complexity:** $O(1)$ (or $O(26)$). The `count` dictionary stores at most 26 uppercase English characters.

## Minimum Window Substring

````python
""" 
https://neetcode.io/problems/minimum-window-with-characters/question?list=blind75

Given two strings s and t, return the shortest substring of s such that every character in t, 
including duplicates, is present in the substring. 
If such a substring does not exist, return an empty string "".

You may assume that the correct output is always unique.

Example 1:

Input: s = "OUZODYXAZV", t = "XYZ"

Output: "YXAZ"
Explanation: "YXAZ" is the shortest substring that includes "X", "Y", and "Z" from string t.

Example 2:

Input: s = "xyz", t = "xyz"

Output: "xyz"
Example 3:

Input: s = "x", t = "xy"

Output: ""
Constraints:

1 <= s.length <= 1000
1 <= t.length <= 1000
s and t consist of uppercase and lowercase English letters.

"""
````

````python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        output = ''
        min_len = float('inf')
        freq = {}
        matched = 0
        left = 0

        for char in t:
            if char not in freq:
                freq[char] = 0
            freq[char] += 1
            
        for right in range(len(s)):
            char = s[right]
            if char in freq:
                freq[char] -= 1
                if freq[char] >= 0:
                	matched += 1
            while matched == len(t):
                if min_len > right - left + 1:
                    min_len = right - left + 1
                    output = s[left:right+1]
                prev_char = s[left]
                if prev_char in freq:
                    freq[prev_char] += 1
                    if freq[prev_char] > 0:
                    	matched -= 1
                left += 1
        return output
    
````

This is the **Minimum Window Substring** problem. It is a hard-level problem often solved using the **Sliding Window** technique with two hash maps (or frequency counters).

### The Logic ($O(N)$)

1.  **Frequency Map:** First, count the characters needed from string `t`. Let's call this `countT`.
2.  **Expand (`r`):** Iterate through `s` with a right pointer `r`. Add the current character to a `window` counter.
3.  **Check Condition:** We need a way to check if the current window has all characters from `t`.
    *   *Optimization:* Instead of comparing two dictionaries every time (which takes $O(26)$ or $O(52)$), we maintain two variables: `have` and `need`.
    *   `need`: Total unique characters in `t`.
    *   `have`: How many unique characters in the current window currently meet the required count.
4.  **Shrink (`l`):** Once `have == need` (meaning the window is valid), we try to make it smaller.
    *   Update the result (start index and length) if the current window is smaller than the previous best.
    *   Remove the character at the left pointer `l` from the `window` count.
    *   If removing that character makes the window invalid (count drops below required), decrement `have`.
    *   Increment `l`.

### Solution in Python

```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        if t == "": return ""
        
        # Dictionary to keep track of the count of characters in t
        countT, window = {}, {}
        for c in t:
            countT[c] = countT.get(c, 0) + 1
            
        have, need = 0, len(countT)
        res, resLen = [-1, -1], float("inf")
        l = 0
        
        for r in range(len(s)):
            c = s[r]
            window[c] = window.get(c, 0) + 1
            
            # If the current character is a required one and 
            # its count matches the requirement in t, increment 'have'
            if c in countT and window[c] == countT[c]:
                have += 1
                
            # While the window is valid (contains all chars from t)
            while have == need:
                # Update our result if this window is smaller
                if (r - l + 1) < resLen:
                    res = [l, r]
                    resLen = (r - l + 1)
                
                # Shrink the window from the left
                window[s[l]] -= 1
                if s[l] in countT and window[s[l]] < countT[s[l]]:
                    have -= 1
                l += 1
                
        l, r = res
        return s[l : r + 1] if resLen != float("inf") else ""
```

### Walkthrough (Example 1: s = "OUZODYXAZV", t = "XYZ")

1.  **Setup:** `countT = {X:1, Y:1, Z:1}`. `need` = 3. `have` = 0.
2.  **Expand:**
    *   `r` moves... eventually window contains "OUZODYX".
    *   Window has X, Y, Z. `have` becomes 3.
3.  **Shrink (Window is valid):**
    *   Current window: "OUZODYX". Length 7. Update `res`.
    *   Remove 'O' (at `l=0`). Valid.
    *   Remove 'U'. Valid.
    *   ...
    *   Eventually `l` moves past 'Z'. Window loses 'Z'. `have` becomes 2. Loop breaks.
4.  **Continue Expanding:**
    *   `r` moves to 'A', then 'Z'.
    *   Current window (roughly): "YXAZ".
    *   `have` becomes 3 again.
5.  **Shrink again:**
    *   Current window: "YXAZ". Length 4. Update `res` (4 is better than 7).
    *   Try to shrink... if we remove 'Y', `have` drops.
6.  **End:** Return "YXAZ".

### Complexity Analysis
*   **Time Complexity:** $O(N)$.
    *   The `r` pointer iterates through `s` once.
    *   The `l` pointer iterates through `s` at most once.
    *   Dictionary operations are $O(1)$.
*   **Space Complexity:** $O(M)$ or $O(1)$.
    *   We need two dictionaries of size at most 52 (uppercase + lowercase English letters), which is constant space relative to $N$.

## Find Minimum in Rotated Sorted Array

````python
""" 
https://neetcode.io/problems/find-minimum-in-rotated-sorted-array/question
You are given an array of length n which was originally sorted in ascending order. 
It has now been rotated between 1 and n times. 
For example, 
the array nums = [1,2,3,4,5,6] might become:

[3,4,5,6,1,2] if it was rotated 4 times.
[1,2,3,4,5,6] if it was rotated 6 times.
Notice that rotating the array 4 times moves the last four elements of the array to the beginning. 
Rotating the array 6 times produces the original array.

Assuming all elements in the rotated sorted array nums are unique, 
return the minimum element of this array.

A solution that runs in O(n) time is trivial, can you write an algorithm that runs in O(log n) time?

Example 1:

Input: nums = [3,4,5,6,1,2]

Output: 1
Example 2:

Input: nums = [4,5,0,1,2,3]

Output: 0
Example 3:

Input: nums = [4,5,6,7]

Output: 4
Constraints:

1 <= nums.length <= 1000
-1000 <= nums[i] <= 1000
"""
````

````python
class Solution:
    def findMin(self, nums: List[int]) -> int:
        l, r = 0, len(nums)-1
        while l < r:
            m = (l + r) // 2
            if nums[m] > nums[r]:
                l = m + 1
            else:
                r = m
        return nums[l]
````

To solve this problem in $O(\log n)$ time, we must use **Binary Search**.

### The Logic

A rotated sorted array can be visualized as two sorted subarrays:
1.  **Left Portion:** Larger values (e.g., `3, 4, 5, 6`)
2.  **Right Portion:** Smaller values (e.g., `1, 2`)

The minimum value is the "pivot" point where the Left Portion ends and the Right Portion begins.

We use binary search with pointers `l` (left) and `r` (right). In every iteration, we compare the middle element `nums[mid]` with the rightmost element `nums[r]`.

1.  **If `nums[mid] > nums[r]`**:
    *   This means `mid` belongs to the **Left Portion** (large values).
    *   The minimum value must be to the **right** of `mid`.
    *   *Action:* `l = mid + 1`

2.  **If `nums[mid] <= nums[r]`**:
    *   This means `mid` belongs to the **Right Portion** (small values), OR the array is currently fully sorted.
    *   The minimum value is either `mid` itself or somewhere to the **left**.
    *   *Action:* `r = mid` (Note: we don't do `mid - 1` because `mid` could be the minimum).

The loop continues until `l` meets `r`, at which point they point to the minimum element.

### Solution

```python
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        l, r = 0, len(nums) - 1
        
        while l < r:
            mid = (l + r) // 2
            
            # Compare middle element with the rightmost element
            if nums[mid] > nums[r]:
                # We are in the left sorted portion (larger numbers).
                # The minimum must be to the right.
                l = mid + 1
            else:
                # We are in the right sorted portion (smaller numbers)
                # or the array is not rotated.
                # The minimum is at 'mid' or to the left.
                r = mid
                
        return nums[l]
```

### Walkthrough (Example: `[3, 4, 5, 1, 2]`)

1.  **Start:** `l=0` (val 3), `r=4` (val 2).
2.  **Iteration 1:**
    *   `mid = 2` (val 5).
    *   Check: Is `5 > 2`? **Yes**.
    *   This implies the dip happens after index 2.
    *   `l` becomes `mid + 1` $\to$ **3**.
3.  **Iteration 2:**
    *   `l=3` (val 1), `r=4` (val 2).
    *   `mid = 3` (val 1).
    *   Check: Is `1 > 2`? **No**.
    *   This implies we are in the smaller sorted part.
    *   `r` becomes `mid` $\to$ **3**.
4.  **End:** `l` (3) is not less than `r` (3). Loop ends.
5.  **Return:** `nums[3]` which is **1**.

### Complexity Analysis
*   **Time Complexity:** $O(\log n)$ because we cut the search space in half during every iteration.
*   **Space Complexity:** $O(1)$ because we only use two pointers.

这是一个非常好的问题。

之所以使用 `l = mid + 1` 而不是 `l = mid`，主要有两个原因：**逻辑上的必然性** 和 **避免死循环**。

### 1. 逻辑原因：`mid` 绝对不可能是最小值

当 `nums[mid] > nums[r]` 条件成立时，意味着 `mid` 处于**左半段（较大值段）**。

举个例子：`[3, 4, 5, 1, 2]`
*   `l=0` (3), `r=4` (2)
*   `mid=2` (值是 5)
*   比较：`5 > 2` 成立。

在这种情况下：
1.  数组是升序旋转的，所以 `nums[mid]` 左边的数（3, 4）都比 `nums[mid]` 小，但也属于“左半段”。
2.  `nums[mid]` 明显比右边的某个数（`nums[r]`）大。

**结论**：既然 `nums[mid]` 比右边某个数大，那么 `nums[mid]` **绝不可能是整个数组的最小值**。
因此，我们可以安全地把 `mid` 排除在搜索范围之外，直接从 `mid + 1` 开始找。

---

### 2. 算法原因：防止死循环

在二分查找中，计算 `mid` 通常是向下取整的：`mid = (l + r) // 2`。

如果只剩两个元素，例如 `[3, 1]`：
*   `l = 0` (值 3)
*   `r = 1` (值 1)
*   `mid = (0 + 1) // 2 = 0` (指向 3)

此时 `nums[mid] > nums[r]` (3 > 1) 成立。

*   **如果你写 `l = mid`**：
    *   `l` 更新为 `0`。
    *   下一轮循环，`l` 还是 0，`r` 还是 1，`mid` 还是 0。
    *   **结果：程序陷入死循环 (Infinite Loop)。**

*   **如果你写 `l = mid + 1`**：
    *   `l` 更新为 `1`。
    *   此时 `l` 等于 `r`，循环结束（`while l < r`），返回 `nums[1]`（即 1），得到正确答案。

### 总结对比

| 情况 | 判断条件 | 处理方式 | 为什么？ |
| :--- | :--- | :--- | :--- |
| **mid 在左半段** | `nums[mid] > nums[r]` | `l = mid + 1` | `mid` 本身肯定不是最小值，最小值在它右边，且为了防止死循环必须 +1。 |
| **mid 在右半段** | `nums[mid] <= nums[r]` | `r = mid` | `mid` **有可能**就是最小值（例如 `[3, 1, 2]` 中间是 1），所以不能 -1，必须保留 `mid`。 |

答案是：**因为 `mid` 自己有可能就是那个最小值。**

如果我们使用 `r = mid - 1`，就会把 `mid` 这个位置从搜索范围内**踢出去**。万一 `mid` 刚好指向的就是最小值，那么我们就永远丢失了正确答案。

让我们通过一个具体例子来看：

### 例子：`[3, 1, 2]`

我们要找最小值 `1`。
*   `l = 0` (值 3)
*   `r = 2` (值 2)
*   **计算 `mid`**： `(0 + 2) // 2 = 1`。此时 `nums[mid]` 是 **1**。

**判断逻辑：**
比较 `nums[mid]` (1) 和 `nums[r]` (2)。
因为 `1 < 2`，进入 `else` 分支。

---

#### 情况 A：如果你写 `r = mid - 1` (错误写法)
*   `r` 变成了 `1 - 1 = 0`。
*   现在的搜索范围 `[l, r]` 变成了 `[0, 0]`，也就是只剩 `[3]` 了。
*   循环结束，你返回了 `3`。
*   **错误！** 最小值明明是 `1`，被你刚才那一步减一给扔掉了。

#### 情况 B：如果你写 `r = mid` (正确写法)
*   `r` 变成了 `1`。
*   现在的搜索范围 `[l, r]` 变成了 `[0, 1]`，包含 `[3, 1]`。
*   最小值 `1` 依然保留在搜索范围内，下一轮循环还能找到它。

---

### 总结核心逻辑

这道题的二分逻辑是不对称的：

1.  **当 `nums[mid] > nums[r]`**：
    *   意味着 `mid` 在左半段（较大值的那一段）。
    *   所以 `mid` **绝对不是**最小值。
    *   我们可以大胆地用 `l = mid + 1` 跳过它。

2.  **当 `nums[mid] <= nums[r]`**：
    *   意味着 `mid` 在右半段（较小值的那一段），或者也就是最小值本身。
    *   例如 `[4, 5, 0, 1, 2]`，`mid` 是 `0`。`0 < 2`。
    *   `mid` **可能是**最小值，也可能最小值还在它左边。
    *   所以我们**必须保留 `mid`**，只能缩减右边界到 `mid`，即 `r = mid`。

**一句话口诀：**
*   比右边大，最小值肯定在右边 (扔掉 `mid`)。
*   比右边小，自己可能是最小值 (保留 `mid`)。

## Search in Rotated Sorted Array

````python
""" 
You are given an array of length n which was originally sorted in ascending order. It has now been rotated between 1 and n times. For example, the array nums = [1,2,3,4,5,6] might become:

[3,4,5,6,1,2] if it was rotated 4 times.
[1,2,3,4,5,6] if it was rotated 6 times.
Given the rotated sorted array nums and an integer target, return the index of target within nums, or -1 if it is not present.

You may assume all elements in the sorted rotated array nums are unique,

A solution that runs in O(n) time is trivial, can you write an algorithm that runs in O(log n) time?

Example 1:

Input: nums = [3,4,5,6,1,2], target = 1

Output: 4
Example 2:

Input: nums = [3,5,6,0,1,2], target = 4

Output: -1
Constraints:

1 <= nums.length <= 1000
-1000 <= nums[i] <= 1000
-1000 <= target <= 1000
All values of nums are unique.
nums is an ascending array that is possibly rotated.

"""
````

````python
from typing import List

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l, r = 0, len(nums) - 1

        while l <= r:
            mid = (l + r) // 2

            if nums[mid] == target:
                return mid

            # Check if the left portion is sorted
            if nums[l] <= nums[mid]:
                # If target is within the sorted left portion
                if nums[l] <= target < nums[mid]:
                    r = mid - 1
                else:
                    l = mid + 1
            # Otherwise, the right portion is sorted
            else:
                # If target is within the sorted right portion
                if nums[mid] < target <= nums[r]:
                    l = mid + 1
                else:
                    r = mid - 1

        return -1
````

This problem is a modification of standard Binary Search.

### The Logic

In a rotated sorted array, if we pick a middle element, one half of the array will always be **strictly sorted**, and the other half will contain the rotation pivot (unsorted).

We can determine which half is sorted by comparing `nums[l]` (leftmost) and `nums[mid]`.

1.  **Calculate Mid**: `mid = (l + r) // 2`.
2.  **Check Target**: If `nums[mid] == target`, return `mid`.
3.  **Identify Sorted Portion**:
    *   **If `nums[l] <= nums[mid]`**: The **left** half `[l...mid]` is sorted.
        *   We then check if the target lies strictly within this sorted range: `nums[l] <= target < nums[mid]`.
        *   If yes, search the left (`r = mid - 1`).
        *   If no, search the right (`l = mid + 1`).
    *   **Else (`nums[l] > nums[mid]`)**: The **right** half `[mid...r]` is sorted.
        *   We check if the target lies strictly within this sorted range: `nums[mid] < target <= nums[r]`.
        *   If yes, search the right (`l = mid + 1`).
        *   If no, search the left (`r = mid - 1`).

### Solution

```python
from typing import List

```

### Walkthrough (Example: `nums = [4,5,6,7,0,1,2]`, `target = 0`)

1.  `l=0`, `r=6`, `mid=3` (`val=7`).
2.  **Is Left Sorted?** `nums[0] (4) <= nums[3] (7)` is **True**.
    *   Is target `0` inside `[4, 7)`? **No**.
    *   Go Right: `l = mid + 1` $\to$ `l = 4`.
3.  `l=4`, `r=6`, `mid=5` (`val=1`).
4.  **Is Left Sorted?** `nums[4] (0) <= nums[5] (1)` is **True**.
    *   Is target `0` inside `[0, 1)`? **Yes**.
    *   Go Left: `r = mid - 1` $\to$ `r = 4`.
5.  `l=4`, `r=4`, `mid=4` (`val=0`).
6.  `nums[mid] == target`. **Return 4**.

### Complexity Analysis
*   **Time Complexity:** $O(\log n)$. In every step, we eliminate half of the array, regardless of whether the pivot exists in that half or not.
*   **Space Complexity:** $O(1)$. We only use a constant amount of extra variables.

## Invert Binary Tree

````python
""" 
https://neetcode.io/problems/invert-a-binary-tree/question?list=blind75
You are given the root of a binary tree root. Invert the binary tree and return its root.

"""
````

````python
from typing import Optional

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:
            return
        if not root.left and not root.right:
            return root
        root.left, root.right = root.right, root.left
        self.invertTree(root.right)
        self.invertTree(root.left)
        return root

        
````

You are given the root of a binary tree root. Invert the binary tree and return its root.

Example 1:

Input: root = [4,2,7,1,3,6,9]

Output: [4,7,2,9,6,3,1]
Example 2:

Input: root = [2,1,3]

Output: [2,3,1]
Constraints:

The number of nodes in the tree is in the range [0, 100].
-100 <= Node.val <= 100

### Approach 1: Recursive DFS (Most Common)

The most intuitive way to solve this is using recursion. For every node, we swap its left and right children, and then recursively do the same for the children's children.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:
            return None
        
        # Swap the children
        root.left, root.right = root.right, root.left
        
        # Recursively invert the subtrees
        self.invertTree(root.left)
        self.invertTree(root.right)
        
        return root
```

*   **Time Complexity:** $O(n)$, where $n$ is the number of nodes. We visit each node exactly once.
*   **Space Complexity:** $O(h)$, where $h$ is the height of the tree (due to the recursion stack). In the worst case (skewed tree), this is $O(n)$.

---

### Approach 2: Iterative BFS (Queue)

If you want to avoid recursion (or are worried about stack overflow on extremely deep trees), you can use a Queue (Breadth-First Search) to process nodes layer by layer.

```python
from collections import deque

class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:
            return None
        
        queue = deque([root])
        
        while queue:
            node = queue.popleft()
            
            # Swap the children
            node.left, node.right = node.right, node.left
            
            # Add children to the queue if they exist
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
                
        return root
```

*   **Time Complexity:** $O(n)$.
*   **Space Complexity:** $O(n)$ in the worst case (the queue might store up to the entire bottom level of the tree).
