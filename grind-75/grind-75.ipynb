## 121. Best Time to Buy and Sell Stock

````python
"""
https://www.techinterviewhandbook.org/grind75/?grouping=topics

https://leetcode.com/problems/best-time-to-buy-and-sell-stock/description/

You are given an array prices where prices[i] is the price of a given stock on the ith day.
You want to maximize your profit by choosing a single day to buy one stock 
and choosing a different day in the future to sell that stock.
Return the maximum profit you can achieve from this transaction. 
If you cannot achieve any profit, return 0. 

Example 1:

Input: prices = [7,1,5,3,6,4]
Output: 5
Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.
Note that buying on day 2 and selling on day 1 is not allowed because you must buy before you sell.
Example 2:

Input: prices = [7,6,4,3,1]
Output: 0
Explanation: In this case, no transactions are done and the max profit = 0.
"""
````

````python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        max_profit = float('-inf')
        min_price = float('inf')
        for price in prices:
            min_price = min(min_price, price)
            max_profit = max(max_profit, price - min_price)
        return max_profit
````

这是一个非常经典的动态规划/贪心算法题目（LeetCode 121. Best Time to Buy and Sell Stock）。

### 解题思路：一次遍历 (One Pass)

我们不需要比较每一天和它之后所有日子的差值（那样是 $O(N^2)$）。我们只需要遍历一次数组，在遍历的过程中记录两个状态：
1.  **历史最低价格 (`min_price`)**：代表截止到当前为止，买入成本最低的那一天。
2.  **当前最大利润 (`max_profit`)**：代表如果在历史最低点买入，并在当前卖出，是否能打破之前的利润记录。

**逻辑步骤：**
*   初始化 `min_price` 为无穷大。
*   初始化 `max_profit` 为 0。
*   遍历每一天的价格 `price`：
    *   **情况 A**：如果 `price` 比 `min_price` 还低，说明我们要更新买入点（因为越低买入越好），更新 `min_price = price`。
    *   **情况 B**：如果 `price` 比 `min_price` 高，说明今天卖出能赚钱。计算 `price - min_price`，如果这个值比 `max_profit` 大，就更新 `max_profit`。

### Python 代码实现

```python
from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        # 如果数组为空或只有一个元素，无法进行交易，返回0
        if not prices:
            return 0
        
        # 初始化最低价格为正无穷大，最大利润为0
        min_price = float('inf')
        max_profit = 0
        
        for price in prices:
            # 1. 寻找历史最低价（假设这一天买入）
            if price < min_price:
                min_price = price
            
            # 2. 如果当前价格大于最低价，计算利润并更新最大值
            # 注意：这里的 elif 保证了我们不会在更新最低价的同一天卖出
            # (虽然同一天买卖利润为0，不影响结果，但逻辑上这样更清晰)
            elif price - min_price > max_profit:
                max_profit = price - min_price
                
        return max_profit

# --- 测试用例 ---
if __name__ == "__main__":
    sol = Solution()
    
    # 示例 1
    prices1 = [7, 1, 5, 3, 6, 4]
    print(f"Input: {prices1}")
    print(f"Output: {sol.maxProfit(prices1)}") 
    # 解释: 在第 2 天（价格=1）买入，在第 5 天（价格=6）卖出，利润 = 6-1 = 5。
    
    print("-" * 20)
    
    # 示例 2
    prices2 = [7, 6, 4, 3, 1]
    print(f"Input: {prices2}")
    print(f"Output: {sol.maxProfit(prices2)}") 
    # 解释: 价格持续下跌，找不到比买入价更高的卖出价，利润为 0。
```

### 复杂度分析

*   **时间复杂度**: $O(N)$
    *   只需要遍历 `prices` 数组一次。
*   **空间复杂度**: $O(1)$
    *   只使用了两个变量 `min_price` 和 `max_profit`。

### 图解推导 (示例 1: `[7, 1, 5, 3, 6, 4]`)

| 当前价格 (Price) | 动作判断 | 更新后的 min_price | 当前计算利润 | 更新后的 max_profit |
| :--- | :--- | :--- | :--- | :--- |
| **7** | 7 < inf (更新最低价) | **7** | - | 0 |

| **4** | 4 > 1 (计算利润) | 1 | 4 - 1 = 3 | 5 (不变) |

最终返回 **5**。

## 169. Majority Element

````python
"""
https://leetcode.com/problems/majority-element/description/

Given an array nums of size n, return the majority element.

The majority element is the element that appears more than ⌊n / 2⌋ times. 
You may assume that the majority element always exists in the array.

 

Example 1:

Input: nums = [3,2,3]
Output: 3
Example 2:

Input: nums = [2,2,1,1,1,2,2]
Output: 2

"""

````

````python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        dict_num = dict()
        for num in nums:
            dict_num[num] = dict_num.get(num, 0) + 1
        max_fre = 0
        max_fre_num = None
        for num, fre in dict_num.items():
            if fre > max_fre:
                max_fre = fre
                max_fre_num = num
        return max_fre_num

        
````

这个问题有几种常见的解法。以下是三种主要的 Python 解法，按推荐程度排序：

### 方法一：Boyer-Moore 摩尔投票算法 (最优解)

这是面试中最希望看到的解法。它的核心思想是“抵消”：将众数视为 +1，非众数视为 -1。因为众数超过总数的一半，所以全部抵消后，剩下的那个一定是众数。

*   **时间复杂度**: $O(n)$
*   **空间复杂度**: $O(1)$

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        count = 0
        candidate = None

        for num in nums:
            # 如果计数器为0，说明之前的都抵消了，重新假设当前数字为众数
            if count == 0:
                candidate = num
            
            # 如果当前数字等于候选众数，票数+1，否则-1
            if num == candidate:
                count += 1
            else:
                count -= 1
        
        return candidate
```

### 方法二：排序法 (代码最简洁)

因为众数出现的次数超过 $n/2$，所以在将数组排序后，众数一定位于数组的中间位置 (下标 $n // 2$)。

*   **时间复杂度**: $O(n \log n)$ (因为排序)
*   **空间复杂度**: $O(1)$ 或 $O(n)$ (取决于排序算法的实现)

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        nums.sort()
        return nums[len(nums) // 2]
```

### 方法三：哈希表 (字典统计)

使用字典统计每个数字出现的次数，一旦发现某个数字次数超过一半直接返回。

*   **时间复杂度**: $O(n)$
*   **空间复杂度**: $O(n)$ (需要存储哈希表)

```python
from collections import Counter

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        counts = Counter(nums)
        # 也可以直接返回出现次数最多的那个
        # return counts.most_common(1)[0][0]
        
        n = len(nums)
        for key, value in counts.items():
            if value > n // 2:
                return key
```

### 总结

*   如果你追求**空间效率**或是在**面试**中，请使用 **摩尔投票法 (方法一)**。
*   如果你想**快速写出代码**且不介意 $O(n \log n)$ 的时间复杂度，使用 **排序法 (方法二)**。
*   这是 LeetCode 第 169 题。

````python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        dict_num = dict()
        majority = len(nums) // 2
        for num in nums:
            dict_num[num] = dict_num.get(num, 0) + 1
            if dict_num[num] > majority:
                return num
        return None
````

## 217. Contains Duplicate

````python
"""
https://leetcode.com/problems/contains-duplicate/description/

Given an integer array nums, return true if any value appears at least twice in the array, 
and return false if every element is distinct. 

Example 1:

Input: nums = [1,2,3,1]

Output: true

Explanation:

The element 1 occurs at the indices 0 and 3.

Example 2:

Input: nums = [1,2,3,4]

Output: false

Explanation:

All elements are distinct.

Example 3:

Input: nums = [1,1,1,3,3,4,3,2,4,2]

Output: true

 

Constraints:

1 <= nums.length <= 105
-109 <= nums[i] <= 109
"""

````

````python
def containsDuplicate(self, nums):
    """
    :type nums: List[int]
    :rtype: bool
    """
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False
````

针对这个问题，有三种常见的 Python 解法。根据面试要求和实际应用场景，推荐程度从高到低排列：

### 方法一：利用集合 (Set) 的特性 (最 Pythonic 的写法)

这是最简洁、最常用的 Python 解法。`set`（集合）是一种自动去重的数据结构。如果将列表转换为集合后，长度变短了，说明原列表中有重复元素。

*   **时间复杂度**: $O(n)$ - 构建集合需要遍历列表。
*   **空间复杂度**: $O(n)$ - 集合需要存储元素。

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        # 如果去重后的长度 不等于 原长度，说明有重复
        return len(set(nums)) != len(nums)
```

---

### 方法二：哈希表/哈希集合 (提前退出)

虽然方法一代码简短，但它必须遍历完整个数组。方法二在面试中常被提及，因为它可以在发现第一个重复元素时**立即停止**，在某些情况下（例如重复元素在数组开头）效率更高。

*   **时间复杂度**: $O(n)$
*   **空间复杂度**: $O(n)$

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()
        for num in nums:
            # 如果当前数字已经在集合里，说明重复了
            if num in seen:
                return True
            seen.add(num)
        return False
```

---

### 方法三：排序法 (节省空间)

先将数组排序，重复的元素一定会相邻。这通常用于**不能使用额外空间**（或者内存非常有限）的场景。

*   **时间复杂度**: $O(n \log n)$ - 排序通常是主要的时间消耗。
*   **空间复杂度**: $O(1)$ 或 $O(n)$ - 取决于 Python `sort()` 的具体实现（通常被认为是 $O(n)$ 因为 Timsort 需要临时空间，但在某些语言中这被视为 $O(1)$ 的解法思路）。

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        nums.sort()
        # 遍历数组，比较当前元素和下一个元素
        for i in range(len(nums) - 1):
            if nums[i] == nums[i+1]:
                return True
        return False
```

### 总结

*   **平时写代码 / LeetCode 刷题**: 推荐 **方法一**（一行代码解决）。
*   **面试中**: 推荐先写方法一，如果面试官问“能不能如果发现重复就立刻停止？”，则写 **方法二**。
*   这是 LeetCode 第 217 题。

## 57. Insert Interval

````python
"""
https://leetcode.com/problems/insert-interval/description/

You are given an array of non-overlapping intervals intervals 
where intervals[i] = [starti, endi] represent the start and the end of the ith interval 
and intervals is sorted in ascending order by starti. 
You are also given an interval newInterval = [start, end] 
that represents the start and end of another interval.

Insert newInterval into intervals such that intervals is still sorted 
in ascending order by starti and intervals still does not have any 
overlapping intervals (merge overlapping intervals if necessary).

Return intervals after the insertion.
Note that you don't need to modify intervals in-place. You can make a new array and return it.

Example 1:
Input: intervals = [[1,3],[6,9]], newInterval = [2,5]
Output: [[1,5],[6,9]]
Example 2:

Input: intervals = [[1,2],[3,5],[6,7],[8,10],[12,16]], newInterval = [4,8]
Output: [[1,2],[3,10],[12,16]]
Explanation: Because the new interval [4,8] overlaps with [3,5],[6,7],[8,10].
 

Constraints:

0 <= intervals.length <= 104
intervals[i].length == 2
0 <= starti <= endi <= 105
intervals is sorted by starti in ascending order.
newInterval.length == 2
0 <= start <= end <= 105

Docstring for grokking-the-coding.grind-75.grind-75.ipynb
"""

````

````python
class Solution(object):
    def insert(self, intervals, newInterval):
        """
        :type intervals: List[List[int]]
        :type newInterval: List[int]
        :rtype: List[List[int]]
        """
        result = []
        n = len(intervals)
        i = 0

        while i < n and intervals[i][1] < newInterval[0]:
            result.append(intervals[i])
            i += 1
        
        while i < n and intervals[i][0] <= newInterval[1]:
            newInterval[0] = min(intervals[i][0], newInterval[0])
            newInterval[1] = max(intervals[i][1], newInterval[1])
            i += 1

        result.append(newInterval)

        while i < n:
            result.append(intervals[i])
            i += 1

        return result
        
````

这道题是 LeetCode 第 57 题 **"Insert Interval" (插入区间)**。

核心思路是：由于原数组是**有序且无重叠**的，我们可以通过一次遍历（线性扫描）将所有区间分为三部分处理：
1.  **左侧不重叠部分**：在 `newInterval` 之前结束的区间，直接加入结果。
2.  **重叠合并部分**：与 `newInterval` 有重叠的区间，不断更新 `newInterval` 的起点和终点，直到合并完成。
3.  **右侧不重叠部分**：在 `newInterval` 之后开始的区间，直接加入结果。

以下是两种 Python 写法：

---

### 方法一：分段处理 (逻辑最清晰)

这种写法使用 `while` 循环分三个阶段处理，逻辑非常符合直觉，面试时最容易写对且不易出错。

*   **时间复杂度**: $O(n)$ - 只需要遍历一次数组。
*   **空间复杂度**: $O(n)$ - 需要创建一个新的列表来存储结果。

```python
class Solution:
    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
        result = []
        i = 0
        n = len(intervals)
        
        # 1. 左侧：将所有在 newInterval 之前结束的区间加入结果
        # intervals[i][1] < newInterval[0] 表示当前区间终点小于新区间起点
        while i < n and intervals[i][1] < newInterval[0]:
            result.append(intervals[i])
            i += 1
            
        # 2. 重叠：如果当前区间与 newInterval 有重叠，合并它们
        # 重叠条件：当前区间起点 <= 新区间终点
        while i < n and intervals[i][0] <= newInterval[1]:
            # 更新 newInterval 的范围：起点取最小，终点取最大
            newInterval[0] = min(newInterval[0], intervals[i][0])
            newInterval[1] = max(newInterval[1], intervals[i][1])
            i += 1
        
        # 将合并后的 newInterval 加入结果
        result.append(newInterval)
        
        # 3. 右侧：将剩余的区间（都在 newInterval 之后）加入结果
        while i < n:
            result.append(intervals[i])
            i += 1
            
        return result
```

---

### 方法二：单次循环 + 切片 (代码更简练)

这个方法利用了 Python 的列表切片功能。如果在遍历过程中发现当前区间完全在 `newInterval` 的右边，说明合并已经结束，可以直接把 `newInterval` 和**剩余的所有区间**一次性返回。

```python
class Solution:
    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
        result = []
        
        for i, interval in enumerate(intervals):
            # 情况1: 当前区间在 newInterval 的左边 (无重叠)
            if interval[1] < newInterval[0]:
                result.append(interval)
            
            # 情况2: 当前区间在 newInterval 的右边 (无重叠)
            elif interval[0] > newInterval[1]:
                # 此时 newInterval 已经确定合并完毕，先加入它
                result.append(newInterval)
                # 再把当前及剩余的所有区间加上，并直接返回
                return result + intervals[i:]
            
            # 情况3: 有重叠
            else:
                # 不断更新 newInterval，但不立即加入 result
                newInterval[0] = min(newInterval[0], interval[0])
                newInterval[1] = max(newInterval[1], interval[1])
        
        # 如果循环结束还没返回，说明 newInterval 是最后一段，或者数组为空
        result.append(newInterval)
        return result
```

### 关键点解析

1.  **重叠的判断**：两个区间 `[a, b]` 和 `[c, d]` 重叠的充要条件是 `a <= d` 且 `c <= b`。
2.  **合并策略**：合并后的区间起点是 `min(start1, start2)`，终点是 `max(end1, end2)`。
3.  **为什么是 O(n)**：虽然看起来我们在操作数组，但我们只是从头到尾扫了一遍。注意方法二中的 `intervals[i:]` 切片操作本身需要复制剩余元素，但因为它是最后一步直接返回，总的操作次数仍然是线性的。

## 15. 3Sum

````python
"""
https://leetcode.com/problems/3sum/description/

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] 
such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation: 
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
Notice that the order of the output and the order of the triplets does not matter.
Example 2:

Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.
Example 3:

Input: nums = [0,0,0]
Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.
 

Constraints:

3 <= nums.length <= 3000
-105 <= nums[i] <= 105
"""
````

````python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        nums.sort()
        res = []
        for i in range(len(nums) - 2):
            if i > 0 and nums[i] == nums[i-1]:
                continue
            self.twoSum(nums, i, res)
        return res
    
    def twoSum(self, nums, i, res):
        left, right = i+1, len(nums)-1
        while left < right:
            if nums[i] + nums[left] + nums[right] == 0:
                res.append([nums[i], nums[left], nums[right]])
                while left < right and nums[left] == nums[left+1]:
                    left += 1
                while left < right and nums[right] == nums[right-1]:
                    right -= 1
                left += 1
                right -= 1
            elif nums[i] + nums[left] + nums[right] > 0:
                right -= 1
            else:
                left += 1
            
````

这道题是 LeetCode 第 15 题 **"3Sum" (三数之和)**。这是一个非常经典的高频面试题。

### 核心思路：排序 + 双指针

直接暴力求解需要三重循环 $O(N^3)$，会超时。
最优解法是先将数组**排序**，然后固定一个数字，再利用**双指针**寻找另外两个数字，将时间复杂度降低到 $O(N^2)$。

### Python 代码实现

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        # 1. 排序：这是使用双指针的前提，同时也方便去重
        nums.sort()
        res = []
        n = len(nums)
        
        # 遍历数组，固定第一个数 nums[i]
        # 只需要遍历到倒数第三个数，因为需要留两个位置给 left 和 right
        for i in range(n - 2):
            
            # 优化：因为数组有序，如果当前数已经大于0，
            # 后面不可能有负数来抵消，直接结束循环
            if nums[i] > 0:
                break
            
            # 外部去重：如果当前数和前一个数相同，跳过（避免产生重复的三元组）
            if i > 0 and nums[i] == nums[i-1]:
                continue
            
            # 2. 双指针寻找另外两个数
            left = i + 1
            right = n - 1
            
            while left < right:
                total = nums[i] + nums[left] + nums[right]
                
                if total == 0:
                    # 找到一组解
                    res.append([nums[i], nums[left], nums[right]])
                    
                    # 内部去重：找到解后，跳过重复的 left 和 right
                    while left < right and nums[left] == nums[left + 1]:
                        left += 1
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1
                    
                    # 双指针同时收缩
                    left += 1
                    right -= 1
                    
                elif total < 0:
                    # 和太小，左指针右移（让数变大）
                    left += 1
                else:
                    # 和太大，右指针左移（让数变小）
                    right -= 1
                    
        return res
```

### 关键点解析（面试必问）

1.  **为什么要排序？**
    *   排序后，我们可以利用单调性使用双指针，避免暴力搜索。
    *   排序后相同的元素会相邻，极大地简化了**去重**逻辑。

2.  **如何去重？（重点）**
    *   **第一个数的去重**：`if i > 0 and nums[i] == nums[i-1]: continue`。这意味着如果当前的数字 `nums[i]` 和上一次循环用的数字一样，我们就跳过它。
    *   **第二、三个数的去重**：在找到一组解 `total == 0` 后，必须移动 `left` 和 `right` 指针跳过所有相同的值。例如 `[-2, 0, 0, 2, 2]`，如果不跳过，会得到两个 `[-2, 0, 2]`。

3.  **时间复杂度是多少？**
    *   **$O(N^2)$**。
    *   排序耗时 $O(N \log N)$。
    *   外层循环遍历 $N$ 次，内层双指针最坏遍历 $N$ 次，总计 $O(N^2)$。

4.  **空间复杂度是多少？**
    *   **$O(1)$** 或 **$O(N)$**。
    *   这取决于排序算法的具体实现（Python 的 Timsort 需要 $O(N)$ 空间）。如果不算存储结果的空间，且允许修改原数组，通常视为 $O(1)$ 的额外空间逻辑。

## 238. Product of Array Except Self

````python
"""
https://leetcode.com/problems/product-of-array-except-self/description/

Given an integer array nums, 
return an array answer such that answer[i] is equal 
to the product of all the elements of nums except nums[i].

The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.

You must write an algorithm that runs in O(n) time and without using the division operation.

Example 1:

Input: nums = [1,2,3,4]
Output: [24,12,8,6]
Example 2:

Input: nums = [-1,1,0,-3,3]
Output: [0,0,9,0,0]
 
Constraints:

2 <= nums.length <= 105
-30 <= nums[i] <= 30
The input is generated such that answer[i] is guaranteed to fit in a 32-bit integer.
 
Follow up: Can you solve the problem in O(1) extra space complexity? 
(The output array does not count as extra space for space complexity analysis.)

"""
````

````python
class Solution(object):
    def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        n = len(nums)
        res = [1]*n
        left = 1
        for i in range(n):
            res[i] = left
            left *= nums[i] 
        right = 1
        for i in range(n-1, -1, -1):
            res[i] *= right
            right *= nums[i]
        
        return res
````

这道题是 LeetCode 第 238 题 **"Product of Array Except Self" (除自身以外数组的乘积)**。

题目有两个硬性要求：
1.  **时间复杂度必须是 $O(n)$**。
2.  **不能使用除法**（这意味着不能算出总乘积然后除以当前元素）。
3.  **进阶要求**：使用 $O(1)$ 的额外空间（输出数组不算在内）。

### 核心思路：左乘积 × 右乘积

对于数组中的任意一个元素 `nums[i]`，它的结果等于：
**（它左边所有元素的乘积）×（它右边所有元素的乘积）**

我们可以利用**两次遍历**来达成：
1.  **第一次遍历（从左到右）**：计算每个元素**左边**所有数的乘积，存入结果数组。
2.  **第二次遍历（从右到左）**：计算每个元素**右边**所有数的乘积，并乘到刚才的结果数组中。

这样我们既没有使用除法，也只用了 $O(1)$ 的额外空间（除了返回数组本身）。

---

### Python 代码解法 (最优解)

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        # 初始化结果数组，output 数组不计入空间复杂度
        answer = [1] * n
        
        # 1. 第一遍遍历：计算左侧所有元素的乘积
        # answer[i] 目前存储的是 i 左边所有数的乘积
        left_product = 1
        for i in range(n):
            answer[i] = left_product
            left_product *= nums[i]
        
        # 此时 answer 数组已经是 [1, nums[0], nums[0]*nums[1], ...]
        
        # 2. 第二遍遍历：计算右侧所有元素的乘积，并直接乘入 answer
        right_product = 1
        for i in range(n - 1, -1, -1):
            answer[i] *= right_product
            right_product *= nums[i]
            
        return answer
```

### 详细图解 (以 `[1, 2, 3, 4]` 为例)

初始化 `answer = [1, 1, 1, 1]`

**第一轮：从左往右 (计算前缀乘积)**
*   `i=0`: `answer[0] = 1` (左边没数)，`left_product` 变 `1*1=1`
*   `i=1`: `answer[1] = 1`，`left_product` 变 `1*2=2`
*   `i=2`: `answer[2] = 2`，`left_product` 变 `2*3=6`
*   `i=3`: `answer[3] = 6`，`left_product` 变 `6*4=24`
*   **此时 answer**: `[1, 1, 2, 6]` (这就是每个位置左边的乘积)

**第二轮：从右往左 (乘上后缀乘积)**
`right_product` 初始为 1
*   `i=3`: `answer[3] = 6 * 1 = 6`，`right_product` 变 `1*4=4`
*   `i=2`: `answer[2] = 2 * 4 = 8`，`right_product` 变 `4*3=12`
*   `i=1`: `answer[1] = 1 * 12 = 12`，`right_product` 变 `12*2=24`
*   `i=0`: `answer[0] = 1 * 24 = 24`，`right_product` 变 `24*1=24`
*   **最终 answer**: `[24, 12, 8, 6]`

### 复杂度分析

*   **时间复杂度**: $O(n)$。我们遍历了数组两次，由线性时间组成。
*   **空间复杂度**: $O(1)$。虽然我们创建了 `answer` 数组，但题目说明**输出数组不计入空间复杂度分析**。我们在处理过程中只使用了 `left_product` 和 `right_product` 两个额外的整数变量。

## 39. Combination Sum

````python
"""
https://leetcode.com/problems/combination-sum/description/

Given an array of distinct integers candidates and a target integer target, 
return a list of all unique combinations of candidates where the chosen numbers sum to target. 
You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. 
Two combinations are unique if the frequency of at least one of the chosen numbers is different.

The test cases are generated such that the number of unique combinations 
that sum up to target is less than 150 combinations for the given input.
 

Example 1:

Input: candidates = [2,3,6,7], target = 7
Output: [[2,2,3],[7]]
Explanation:
2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
7 is a candidate, and 7 = 7.
These are the only two combinations.
Example 2:

Input: candidates = [2,3,5], target = 8
Output: [[2,2,2,2],[2,3,3],[3,5]]
Example 3:

Input: candidates = [2], target = 1
Output: []
 

Constraints:

1 <= candidates.length <= 30
2 <= candidates[i] <= 40
All elements of candidates are distinct.
1 <= target <= 40

"""
````

````python
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        candidates.sort()
        comb = []
        path = []

        def dfs(nums, target, path):
            if target == 0:
                comb.append(path)
                return
            
            for i in range(len(nums)):
                if nums[i] > target:
                    break
                dfs(nums[i:], target-nums[i], path+[nums[i]])
            
        dfs(candidates, target, path)

        return comb

        
````

你的代码主要有一个**逻辑错误**，导致你无法得到包含重复数字的组合（比如 `[2, 2, 3]`），只能得到所有数字都不重复的组合（类似于 Combination Sum II）。

### 错误原因

问题出在递归调用的这一行：

```python
# 你的代码
dfs(nums[i+1:], target-nums[i], path+[nums[i]])
```

你使用了 `nums[i+1:]` 进行切片。
这意味着：**“在下一层递归中，只能从当前数字的【下一个】数字开始选。”**

*   **题目要求**：数字可以被**无限次重复使用** ("The same number may be chosen from candidates an unlimited number of times")。
*   **你的逻辑**：选了当前这个数后，下次就把这个数从列表里剔除了，再也不能选它了。

**举例说明**：
假设 `candidates = [2, 3]`, `target = 7`。
1.  当前选了 `2`。
2.  你传入下一层的是 `nums[i+1:]` 即 `[3]`。
3.  下一层只能选 `3`，**永远无法再次选到 2**。
4.  所以你永远凑不出 `2 + 2 + 3 = 7`。

---

### 修改方法 1：最小改动 (不推荐，但能跑通)

如果你想保留你现在的切片写法，只需要把 `i+1` 改成 `i`。这样下一层递归还能看到当前这个数字。

```python
# 修改后：传入 nums[i:]，包含了当前数字 nums[i] 以及后面的数字
dfs(nums[i:], target-nums[i], path+[nums[i]])
```

---

### 修改方法 2：引入 start_index (推荐，标准解法)

在面试中，不推荐使用 `nums[i:]` 这种切片方式传递参数，因为：
1.  **空间浪费**：每次递归都复制并创建一个新的列表，空间复杂度很高。
2.  **效率低**：列表切片和列表相加 (`path + [...]`) 都会产生额外的开销。

推荐使用 **下标 (index)** 来控制遍历范围：

```python
class Solution(object):
    def combinationSum(self, candidates, target):
        candidates.sort()
        comb = []
        # path 建议用回溯法复用，不要每次都创建新列表
        path = []

        # 引入 start_index，表示当前可以从 candidates 的哪个位置开始选
        def dfs(target, start_index, current_path):
            if target == 0:
                # 注意：因为 path 是引用的，所以要拷贝一份加入结果
                comb.append(list(current_path)) 
                return
            
            # 从 start_index 开始遍历，保证了：
            # 1. 包含了当前数字 (可以重复选)
            # 2. 不会回头选之前的数字 (去重)
            for i in range(start_index, len(candidates)):
                if candidates[i] > target:
                    break
                
                current_path.append(candidates[i])
                
                # 关键点：这里传入 i，而不是 i + 1
                # 表示下一层还可以继续选 candidates[i]
                dfs(target - candidates[i], i, current_path)
                
                # 回溯：撤销选择
                current_path.pop()
            
        dfs(target, 0, path)

        return comb
```

### 总结
你的代码通过了“不重复数字组合”的逻辑，但题目要求“可重复使用”。
**核心修复**：递归时不要跳过当前数字，要允许下一层继续访问当前索引 `i`。

针对 **Combination Sum (组合总和)** 这道题（允许数字无限重复使用），其复杂度分析比一般的遍历要稍微复杂一些。

假设：
*   $N$ = `len(candidates)` (数组长度)
*   $T$ = `target` (目标值)
*   $M$ = `min(candidates)` (数组中最小的那个数)

---

### 1. 时间复杂度 (Time Complexity)

**结论**：$O(N^{\frac{T}{M}})$ （这是一个宽松的上界）

**详细分析**：
这是一个回溯问题，本质是在遍历一棵决策树。我们需要估算这棵树有多少个节点。

1.  **树的最大深度 (Height)**：
    *   最坏的情况是我们一直选数组里最小的那个数。
    *   例如：`candidates = [2]`, `target = 10`。我们需要选 5 次 ($10/2$)。
    *   所以，递归树的最大深度是 $k = \frac{T}{M}$。

2.  **树的分支数 (Branching Factor)**：
    *   在每个节点，我们理论上最多可以尝试选列表中的 $N$ 个数。
    *   虽然我们在代码中用了 `start_index` 来减少分支（剪枝），但在最坏情况的大O分析中，我们通常考虑最宽的情况。

3.  **总节点数**：
    *   一棵深度为 $k$、分支数为 $N$ 的树，其节点总数大约是 $N^k$。
    *   代入 $k$，即 $O(N^{\frac{T}{M}})$。

4.  **复制结果的开销**：
    *   每当找到一个解时，我们需要 `comb.append(list(current_path))`。复制列表需要 $O(k)$ 的时间。
    *   但这部分开销通常被包含在指数级的节点遍历中，所以在 Big-O 表示法中，主要由树的大小决定。

> **注意**：实际运行时间通常远快于这个上界，因为：
> 1. `candidates.sort()` 后的剪枝 (`break`) 会砍掉大量分支。
> 2. `start_index` 使得树的分支越来越少（组合而非排列）。

---

### 2. 空间复杂度 (Space Complexity)

**结论**：$O(\frac{T}{M})$

**详细分析**：
空间复杂度主要取决于**递归栈的深度**以及**临时路径列表 (`path`) 的长度**。

1.  **递归栈深度**：
    *   和时间复杂度分析一样，递归最深只会达到 $T/M$ 层（全选最小数的情况）。
    *   例如 `target=10, min=2`，栈深为 5。

2.  **临时路径列表 (`path`)**：
    *   `path` 列表中最多也只存 $T/M$ 个数字。

3.  **关于结果数组 (`comb`)**：
    *   在算法复杂度分析中，**通常不计算用于存储最终输出结果的空间**（除非面试官特意问）。
    *   如果算上存储结果的空间，那就是 (解的个数 $\times$ 解的平均长度)。

---

### 总结 (可以直接背这个)

*   **时间复杂度**: $O(N^{\frac{T}{M}})$ —— 其中 $N$ 是数组长度，$T$ 是目标值，$M$ 是数组中的最小值。
*   **空间复杂度**: $O(\frac{T}{M})$ —— 主要用于递归栈和当前路径的存储（不包含输出结果）。

## 56. Merge Intervals

````python
"""
https://leetcode.com/problems/merge-intervals/description/

Given an array of intervals where intervals[i] = [starti, endi], 
merge all overlapping intervals, 
and return an array of the non-overlapping intervals that cover all the intervals in the input.

Example 1:

Input: intervals = [[1,3],[2,6],[8,10],[15,18]]
Output: [[1,6],[8,10],[15,18]]
Explanation: Since intervals [1,3] and [2,6] overlap, merge them into [1,6].
Example 2:

Input: intervals = [[1,4],[4,5]]
Output: [[1,5]]
Explanation: Intervals [1,4] and [4,5] are considered overlapping.
Example 3:

Input: intervals = [[4,7],[1,4]]
Output: [[1,7]]
Explanation: Intervals [1,4] and [4,7] are considered overlapping.
 

Constraints:

1 <= intervals.length <= 104
intervals[i].length == 2
0 <= starti <= endi <= 104

"""

````

````python
class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """
        merged = []
        intervals.sort(key=lambda x:x[0])
        for interval in intervals:
            if not merged or merged[-1][1] < interval[0]:
                merged.append(interval)
            else:
                merged[-1][1] = max(interval[1], merged[-1][1])
        return merged
        
````

这道题是 LeetCode 第 56 题 **"Merge Intervals" (合并区间)**。这是一道非常经典的数组/排序类题目，面试中出现频率极高。

### 核心思路：排序 + 一次遍历

如果不排序，我们很难判断哪些区间是相邻或重叠的。
因此，解题的第一步必须是 **按照区间的起点（start）进行排序**。

排序后，所有能合并的区间在数组中一定是**连续**的。我们只需要遍历一遍数组，比较“当前区间”和“结果集中最后一个区间”是否有重叠即可。

#### 算法步骤：

1.  **排序**：将 `intervals` 按照每个区间的**起点** (`x[0]`) 从小到大排序。
2.  **初始化**：创建一个 `merged` 列表用于存储结果。
3.  **遍历**：
    *   如果 `merged` 为空，或者 **当前区间的起点 > `merged` 最后一个区间的终点**：
        *   说明没有重叠，直接将当前区间加入 `merged`。
    *   否则（说明有重叠）：
        *   **合并区间**：我们要更新 `merged` 中最后一个区间的终点。
        *   新的终点取最大值：`max(原终点, 当前区间的终点)`。

---

### Python 代码实现

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # 1. 按照区间的起点进行排序
        intervals.sort(key=lambda x: x[0])
        
        merged = []
        
        for interval in intervals:
            # 如果结果列表为空，或者当前区间与前一个区间不重叠
            if not merged or merged[-1][1] < interval[0]:
                merged.append(interval)
            else:
                # 发生重叠，合并区间
                # 更新前一个区间的终点为两者的最大终点
                merged[-1][1] = max(merged[-1][1], interval[1])
                
        return merged
```

### 详细图解 (以 `[[1,3],[2,6],[8,10],[15,18]]` 为例)

1.  **排序**：`[[1,3], [2,6], [8,10], [15,18]]` (已经是排好序的)。
2.  **遍历**：
    *   `[1,3]`：`merged` 为空 -> 加入 `[1,3]`。`merged = [[1,3]]`。
    *   `[2,6]`：起点 `2` <= `merged` 最后终点 `3` -> **重叠**。
        *   更新终点：`max(3, 6) = 6`。
        *   `merged` 变为 `[[1,6]]`。
    *   `[8,10]`：起点 `8` > `merged` 最后终点 `6` -> **不重叠**。
        *   加入。`merged = [[1,6], [8,10]]`。
    *   `[15,18]`：起点 `15` > `merged` 最后终点 `10` -> **不重叠**。
        *   加入。`merged = [[1,6], [8,10], [15,18]]`。
3.  **返回结果**。

### 复杂度分析

*   **时间复杂度**: $O(N \log N)$
    *   主要消耗在排序上，Python 的 Timsort 复杂度为 $O(N \log N)$。
    *   遍历数组只需要 $O(N)$。
    *   总体复杂度由排序决定。
*   **空间复杂度**: $O(N)$ (或者 $O(\log N)$)
    *   需要 $O(N)$ 的空间来存储结果数组 `merged`。
    *   排序本身通常需要 $O(\log N)$ 到 $O(N)$ 的栈空间（取决于具体语言实现）。

### 常见边界情况

1.  **完全包含**：例如 `[[1, 10], [2, 3]]`。
    *   排序后：`[1, 10]` 先进。
    *   处理 `[2, 3]`：`2 <= 10` 重叠。更新终点 `max(10, 3) = 10`。
    *   结果：`[[1, 10]]`。逻辑正确。
2.  **刚好接触**：例如 `[[1, 4], [4, 5]]`。
    *   `4 <= 4` 重叠。更新终点 `max(4, 5) = 5`。
    *   结果：`[[1, 5]]`。逻辑正确。
3.  **乱序输入**：例如 `[[4,7],[1,4]]`。
    *   因为第一步做了 `sort`，会变成 `[[1,4], [4,7]]`，然后正常合并。逻辑正确。

## 75. Sort Colors

````python
"""
https://leetcode.com/problems/sort-colors/description/

Given an array nums with n objects colored red, white, or blue, 
sort them in-place so that objects of the same color are adjacent, with the colors in the order red, white, and blue.
We will use the integers 0, 1, and 2 to represent the color red, white, and blue, respectively.
You must solve this problem without using the library's sort function.

Example 1:

Input: nums = [2,0,2,1,1,0]
Output: [0,0,1,1,2,2]
Example 2:

Input: nums = [2,0,1]
Output: [0,1,2]
 

Constraints:
n == nums.length
1 <= n <= 300
nums[i] is either 0, 1, or 2.
 

Follow up: Could you come up with a one-pass algorithm using only constant extra space?

"""
````

````python
class Solution(object):
    def sortColors(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        curr = 0

        for i in range(len(nums)):
            if nums[i] == 0:
                nums[i], nums[curr] = nums[curr], nums[i]
                curr += 1
        
        for i in range(len(nums)):
            if nums[i] == 1:
                nums[i], nums[curr] = nums[curr], nums[i]
                curr += 1
        
        return nums
            
        
````

除了经典的**三指针（荷兰国旗）**和**计数排序**外，确实还有几种有趣的思路。虽然它们可能在时间或空间上不如三指针法完美，但能体现不同的算法思维。

### 方法三：两次遍历 - 分别处理法 (单向指针)

如果你觉得三个指针同时动太容易晕，可以把它拆解成**两次简单的交换**。
1.  第一次遍历：先把所有的 `0` 交换到数组的最前面。
2.  第二次遍历：从处理完 `0` 的位置开始，把所有的 `1` 交换到前面。
剩下的自然就是 `2` 了。

*   **优点**：逻辑非常简单，不容易写错边界条件。
*   **缺点**：遍历了两次数组（2N），虽然还是 $O(N)$，但比三指针略慢一点点。

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        n = len(nums)
        ptr = 0 # 指向头部，用于存放当前要处理的数字
        
        # 第一轮：把所有的 0 移到前面
        for i in range(n):
            if nums[i] == 0:
                nums[ptr], nums[i] = nums[i], nums[ptr]
                ptr += 1
        
        # 第二轮：从 ptr 开始，把所有的 1 移到前面
        # 此时 ptr 之前的都是 0，不需要动
        for i in range(ptr, n):
            if nums[i] == 1:
                nums[ptr], nums[i] = nums[i], nums[ptr]
                ptr += 1
```

### 方法四：覆盖法 (极简代码)

这是一个非常巧妙的**一趟扫描**解法，利用了“覆盖”的思想。
我们维护两个指针 `p0` 和 `p1`，分别代表“0的右边界”和“1的右边界”。
当我们遍历数组时：
1.  无论当前数字是几，先假设它是 `2`，填入 `nums[i]`。
2.  如果原数字小于 2（是 0 或 1），说明刚才那个 `2` 填早了，我们把它改成 `1`，并将 `1` 的边界 `p1` 向右移。
3.  如果原数字是 0，说明刚才那个 `1` 也填早了，我们把它改成 `0`，并将 `0` 的边界 `p0` 向右移。

*   **特点**：代码极短，只需要一次循环。
*   **难点**：理解逻辑比较费劲，它是通过“后盖前”的方式自动修正数值。

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        p0 = 0 # 0 的右边界
        p1 = 0 # 1 的右边界
        
        for i in range(len(nums)):
            val = nums[i] # 暂存当前值，因为下面会覆盖 nums[i]
            
            # 第一步：不管三七二十一，先赋值为 2
            nums[i] = 2 
            
            # 第二步：如果原值是 1 或 0，说明刚才把 1 的位置占了，要补回来
            if val < 2:
                nums[p1] = 1
                p1 += 1
            
            # 第三步：如果原值是 0，说明刚才把 0 的位置占了，要补回来
            # 注意：这里会覆盖掉上面写的 1，这是正确的，因为 0 排在 1 前面
            if val == 0:
                nums[p0] = 0
                p0 += 1
```

### 方法五：通用的排序算法 (手写)

虽然题目禁止直接调用 `sort()` 库函数，但你可以**手写**一个标准的排序算法，比如冒泡排序、插入排序或快速排序。
这通常不是面试官想要的（因为没有利用“只有三个数字”这个特性），但它是兜底的解法。

**比如手写一个插入排序 (Insertion Sort):**

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        for i in range(1, len(nums)):
            key = nums[i]
            j = i - 1
            while j >= 0 and key < nums[j]:
                nums[j + 1] = nums[j]
                j -= 1
            nums[j + 1] = key
```

1.  **首选**：**三指针法 (荷兰国旗)**。这是最标准、考察最频繁的解法，不仅一趟扫描，而且只交换必要的元素。
2.  **次选**：**计数排序**。最容易想到，且代码不容易出错。
3.  **备选**：**两次遍历法**。如果你在面试中对三指针的 `curr` 到底要不要 `+1` 感到混乱，用这个方法能稳妥地解决问题。

## 11. Container With Most Water

````python
"""
https://leetcode.com/problems/container-with-most-water/description/

You are given an integer array height of length n. 
There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.
"""

````

````python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        left, right = 0, len(height) - 1
        max_area = 0
        while left < right:
            max_area = max(max_area, (right - left)*min(height[right], height[left]))
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_area
    
````

这道题是 LeetCode 第 11 题 **"Container With Most Water" (盛最多水的容器)**。这是一道经典的**双指针**面试题。

### 核心思路：双指针法 (Two Pointers)

我们可以把这个问题看作是寻找两个柱子，使得它们围成的矩形面积最大。
面积公式：$Area = (j - i) \times \min(height[i], height[j])$

**策略：**
1.  **初始状态**：我们从**最宽**的容器开始，即指针分别指向数组的最左端 (`left = 0`) 和最右端 (`right = n - 1`)。
2.  **移动规则**：
    *   计算当前的面积，更新最大值。
    *   **哪边的柱子比较短，就移动哪边的指针**。
    *   为什么要移动短的？因为容器的高度取决于短板。如果移动长板，宽度变小，而高度不可能变高（受限于短板），面积一定变小。只有移动短板，才有可能找到更高的柱子，从而弥补宽度的损失，使面积变大。
3.  **终止条件**：当 `left` 和 `right` 相遇时结束。

---

### Python 代码实现

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        # 初始化双指针
        left = 0
        right = len(height) - 1
        max_val = 0
        
        while left < right:
            # 计算当前的宽高
            w = right - left
            h = min(height[left], height[right])
            
            # 更新最大面积
            max_val = max(max_val, w * h)
            
            # 贪心策略：移动较短的那根柱子
            # 因为受限于短板，如果不移动短板而移动长板，
            # 宽度减小且高度最多只能等于原来的短板，面积只会变小。
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
                
        return max_val
```

### 详细图解 (以 `height = [1,8,6,2,5,4,8,3,7]` 为例)

1.  **Start**: `left=0` (val=1), `right=8` (val=7).
    *   宽=8, 高=min(1,7)=1. 面积=8.
    *   `height[left]` (1) < `height[right]` (7)，**left 右移**。

2.  **Next**: `left=1` (val=8), `right=8` (val=7).
    *   宽=7, 高=min(8,7)=7. 面积=49. (**更新最大值**)
    *   `height[right]` (7) < `height[left]` (8)，**right 左移**。

3.  **Next**: `left=1` (val=8), `right=7` (val=3).
    *   宽=6, 高=min(8,3)=3. 面积=18.
    *   `height[right]` (3) < `height[left]` (8)，**right 左移**。

4.  **Next**: `left=1` (val=8), `right=6` (val=8).
    *   宽=5, 高=min(8,8)=8. 面积=40.
    *   两者相等，移动任意一个即可（代码中是 `else` 移动 right）。

...以此类推，直到指针相遇。

### 复杂度分析

*   **时间复杂度**: $O(N)$。双指针总共遍历数组一次。
*   **空间复杂度**: $O(1)$。只需要常数级变量存储指针和结果。

### 面试提示
这道题不需要动态规划，也不需要暴力破解 ($O(N^2)$ 会超时)。核心在于解释清楚**“为什么每次要移动较短的那根柱子”**。简而言之：**保留短板没有前途，只有放弃短板才有一线生机**。

## 20. Valid Parentheses

````python
""" 
https://leetcode.com/problems/valid-parentheses/description/

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', 
determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Every close bracket has a corresponding open bracket of the same type.
 
Example 1:

Input: s = "()"

Output: true

Example 2:

Input: s = "()[]{}"

Output: true

Example 3:

Input: s = "(]"

Output: false

Example 4:

Input: s = "([])"

Output: true

Example 5:

Input: s = "([)]"

Output: false

Constraints:

1 <= s.length <= 104
s consists of parentheses only '()[]{}'.

"""
````

````python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        parenthese = {')':'(', '}':'{', ']':'['}
        stack = []

        for ch in s:
            if ch in ['(', '{', '[']:
                stack.append(ch)
            elif ch in parenthese:
                if stack and parenthese[ch] == stack[-1]:
                    stack.pop()
                else:
                    return False
        
        return not stack
        
````

这道题是 LeetCode 第 20 题 **"Valid Parentheses" (有效的括号)**。这是栈 (Stack) 数据结构最经典的入门题。

### 核心思路：栈 (Stack)

括号匹配遵循 **"后进先出" (LIFO)** 的原则：最后出现的左括号，必须最先被右括号闭合。这正好符合**栈**的特性。

**算法流程**：
1.  创建一个空栈。
2.  创建一个字典（哈希表）来建立右括号到左括号的映射关系，方便查询。
3.  遍历字符串中的每个字符：
    *   如果是 **左括号** (`(`, `{`, `[`): 将其 **压入栈 (push)**。
    *   如果是 **右括号** (`)`, `}`, `]`):
        *   如果栈是空的（说明没有对应的左括号），直接返回 `False`。
        *   **弹出栈顶元素 (pop)**，判断它是否与当前的右括号匹配。如果不匹配，返回 `False`。
4.  遍历结束后，如果栈是空的，说明所有括号都正确闭合，返回 `True`；如果栈不为空（有剩余的左括号），返回 `False`。

---

### Python 代码实现

```python
class Solution:
    def isValid(self, s: str) -> bool:
        # 优化：如果字符串长度是奇数，肯定无法完全配对，直接返回 False
        if len(s) % 2 != 0:
            return False
        
        # 建立映射表：右括号 -> 左括号
        pairs = {
            ")": "(",
            "]": "[",
            "}": "{"
        }
        
        stack = []
        
        for char in s:
            # 如果当前字符是右括号（即在 keys 中）
            if char in pairs:
                # 检查栈是否为空
                if not stack:
                    return False
                
                # 弹出栈顶元素（最近的一个左括号）
                top_element = stack.pop()
                
                # 检查是否匹配
                if top_element != pairs[char]:
                    return False
            else:
                # 如果是左括号，直接入栈
                stack.append(char)
        
        # 最后检查栈是否为空（防止类似 "(()" 这种情况）
        return not stack
```

### 极简写法 (利用 Python 特性)

我们可以稍微转换一下思路：遇到左括号时，直接把**它期待的右括号**压入栈中。这样遇到右括号时，只需要判断是否和栈顶相等即可。

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        mapping = {'(': ')', '[': ']', '{': '}'}
        
        for char in s:
            if char in mapping:
                # 如果是左括号，把对应的右括号入栈
                stack.append(mapping[char])
            else:
                # 如果是右括号
                # 1. 栈不能为空
                # 2. 栈顶必须等于当前字符
                if not stack or stack.pop() != char:
                    return False
                    
        return not stack
```

### 复杂度分析

*   **时间复杂度**: $O(n)$。我们需要遍历一遍字符串，每个字符入栈和出栈的操作都是 $O(1)$ 的。
*   **空间复杂度**: $O(n)$。在最坏情况下（例如全是左括号 `"((((("`），栈的大小会达到 $n$。

### 常见边界情况
1.  **奇数长度**: `s = "["` -> `False` (代码开头已优化处理)。
2.  **右括号开头**: `s = "]"` -> `False` (在 `if not stack` 处拦截)。
3.  **左括号未闭合**: `s = "(()"` -> `False` (最后 `return not stack` 拦截)。
4.  **嵌套错误**: `s = "([)]"` -> `False` (匹配检查时拦截)。

## 232. Implement Queue using Stacks

````python
""" 
https://leetcode.com/problems/implement-queue-using-stacks/description/

Implement a first in first out (FIFO) queue using only two stacks. 
The implemented queue should support all the functions of a normal queue (push, peek, pop, and empty).

Implement the MyQueue class:

void push(int x) Pushes element x to the back of the queue.
int pop() Removes the element from the front of the queue and returns it.
int peek() Returns the element at the front of the queue.
boolean empty() Returns true if the queue is empty, false otherwise.
Notes:

You must use only standard operations of a stack, which means only push to top, peek/pop from top, size, 
and is empty operations are valid.
Depending on your language, the stack may not be supported natively. 
You may simulate a stack using a list or deque (double-ended queue) 
as long as you use only a stack's standard operations.
 

Example 1:

Input
["MyQueue", "push", "push", "peek", "pop", "empty"]
[[], [1], [2], [], [], []]
Output
[null, null, null, 1, 1, false]

Explanation
MyQueue myQueue = new MyQueue();
myQueue.push(1); // queue is: [1]
myQueue.push(2); // queue is: [1, 2] (leftmost is front of the queue)
myQueue.peek(); // return 1
myQueue.pop(); // return 1, queue is [2]
myQueue.empty(); // return false
 

Constraints:

1 <= x <= 9
At most 100 calls will be made to push, pop, peek, and empty.
All the calls to pop and peek are valid.
 

Follow-up: Can you implement the queue such that each operation is amortized O(1) time complexity? 
In other words, performing n operations will take overall O(n) time even if one of those operations may take longer.

"""
````

````python
class MyQueue(object):

    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        self.in_stack.append(x)
        
    def pop(self):
        """
        :rtype: int
        """
        if not self.out_stack:
            self.move_in_to_out()
        return self.out_stack.pop()

    def peek(self):
        """
        :rtype: int
        """
        if not self.out_stack:
            self.move_in_to_out()
        return self.out_stack[-1]

    def empty(self):
        """
        :rtype: bool
        """
        return not self.in_stack and not self.out_stack
    
    def move_in_to_out(self):
        while self.in_stack:
            self.out_stack.append(self.in_stack.pop())
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
````

这是一个经典的算法面试题。使用两个栈（Stack）来实现一个队列（Queue）。

### 核心思路

队列是 **FIFO**（先进先出），而栈是 **LIFO**（后进先出）。
我们需要用两个栈来配合：
1.  **输入栈 (`in_stack`)**：只负责接收新的数据（Push）。
2.  **输出栈 (`out_stack`)**：只负责弹出数据或查看队头（Pop/Peek）。

**逻辑如下：**
*   **Push**: 直接把数据压入 `in_stack`。
*   **Pop/Peek**:
    *   如果 `out_stack` 不为空，直接从 `out_stack` 顶部拿数据（因为这里的数据已经是倒序过的，最上面的就是最早进来的）。
    *   如果 `out_stack` 为空，则把 `in_stack` 里的**所有**数据依次弹出并压入 `out_stack`。这就完成了一次“倒序”，`in_stack` 底部的老数据跑到了 `out_stack` 的顶部。

### Python 代码实现

在 Python 中，我们使用 `list` 来模拟栈，只使用 `append()` (push) 和 `pop()` (pop) 以及 `[-1]` (peek) 操作。

```python
class MyQueue:

    def __init__(self):
        # in_stack 用于处理 push 操作
        self.in_stack = []
        # out_stack 用于处理 pop 和 peek 操作
        self.out_stack = []

    def push(self, x: int) -> None:
        """
        将元素 x 推到队列的末尾
        """
        self.in_stack.append(x)

    def pop(self) -> int:
        """
        从队列的开头移除并返回元素
        """
        # 如果输出栈为空，需要先从输入栈“倒腾”数据过来
        if not self.out_stack:
            self.move_in_to_out()
        
        # 弹出输出栈顶部的元素
        return self.out_stack.pop()

    def peek(self) -> int:
        """
        返回队列开头的元素
        """
        # 如果输出栈为空，需要先从输入栈“倒腾”数据过来
        if not self.out_stack:
            self.move_in_to_out()
        
        # 返回输出栈顶部的元素（不移除）
        return self.out_stack[-1]

    def empty(self) -> bool:
        """
        如果队列为空，返回 true；否则，返回 false
        """
        # 两个栈都为空，队列才算空
        return not self.in_stack and not self.out_stack

    # 辅助函数：将数据从输入栈移动到输出栈
    def move_in_to_out(self):
        while self.in_stack:
            self.out_stack.append(self.in_stack.pop())

# 使用示例:
# obj = MyQueue()
# obj.push(1)
# obj.push(2)
# param_2 = obj.peek() # 返回 1
# param_3 = obj.pop()  # 返回 1
# param_4 = obj.empty() # 返回 False
```

### 复杂度分析

这是面试中非常重要的一点，特别是关于 `pop` 和 `peek` 的复杂度。

1.  **Time Complexity (时间复杂度):**
    *   `push`: **$O(1)$**。直接 append 到输入栈。
    *   `empty`: **$O(1)$**。只检查两个列表是否为空。
    *   `pop` 和 `peek`: **摊还复杂度 (Amortized Complexity) 为 $O(1)$**。
        *   **解释**：虽然在最坏情况下（`out_stack` 为空时），我们需要把 `in_stack` 的 $N$ 个元素全部移动一遍，此时是 $O(N)$。
        *   但是，每个元素在其生命周期中，只会被压入 `in_stack` 一次，从 `in_stack` 弹出一次，压入 `out_stack` 一次，从 `out_stack` 弹出一次。平均下来，每个元素的操作次数是常数。因此，平均每次调用的代价是 $O(1)$。

2.  **Space Complexity (空间复杂度):**
    *   **$O(N)$**。其中 $N$ 是队列中的元素数量。我们需要存储所有的元素。

## 150. Evaluate Reverse Polish Notation

````python
"""
https://leetcode.com/problems/evaluate-reverse-polish-notation/description/

You are given an array of strings tokens that represents an arithmetic expression in a Reverse Polish Notation.

Evaluate the expression. Return an integer that represents the value of the expression.

Note that:

The valid operators are '+', '-', '*', and '/'.
Each operand may be an integer or another expression.
The division between two integers always truncates toward zero.
There will not be any division by zero.
The input represents a valid arithmetic expression in a reverse polish notation.
The answer and all the intermediate calculations can be represented in a 32-bit integer.
 

Example 1:

Input: tokens = ["2","1","+","3","*"]
Output: 9
Explanation: ((2 + 1) * 3) = 9
Example 2:

Input: tokens = ["4","13","5","/","+"]
Output: 6
Explanation: (4 + (13 / 5)) = 6
Example 3:

Input: tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
Output: 22
Explanation: ((10 * (6 / ((9 + 3) * -11))) + 17) + 5
= ((10 * (6 / (12 * -11))) + 17) + 5
= ((10 * (6 / -132)) + 17) + 5
= ((10 * 0) + 17) + 5
= (0 + 17) + 5
= 17 + 5
= 22
 

Constraints:

1 <= tokens.length <= 104
tokens[i] is either an operator: "+", "-", "*", or "/", or an integer in the range [-200, 200].

"""

````

````python
class Solution(object):
    def evalRPN(self, tokens):
        """
        :type tokens: List[str]
        :rtype: int
        """
        stack = []

        for token in tokens:
            if token == '+':
                b = stack.pop()
                a = stack.pop()
                stack.append(a + b)
            elif token == '-':
                b = stack.pop()
                a = stack.pop()
                stack.append(a - b)
            elif token == '*':
                b = stack.pop()
                a = stack.pop()
                stack.append(a * b)
            elif token == '/':
                b = stack.pop()
                a = stack.pop()
                stack.append(int(float(a) / b))
            else:
                stack.append(int(token))
        
        return stack[0]

        
````

我们来详细分析一下这道题（逆波兰表达式求值）的时间复杂度和空间复杂度。

假设 `tokens` 数组的长度为 $n$。

### 1. 时间复杂度 (Time Complexity)
**结论：$O(n)$**

*   **遍历过程：** 代码中有一个 `for` 循环，我们需要遍历 `tokens` 列表中的每一个元素，这需要 $n$ 次迭代。
*   **单次操作开销：** 在循环内部，我们要么遇到数字，要么遇到操作符。
    *   **遇到数字：** `stack.append()` 操作是 $O(1)$ 的（在 Python 中列表末尾追加是分摊常数时间）。
    *   **遇到操作符：** 我们执行两次 `stack.pop()`，一次算术运算（`+`, `-`, `*`, `/`），以及一次 `stack.append()`。这些操作也都是 $O(1)$ 的。
    *   **类型转换：** `int()` 和 `float()` 转换对于题目限制的 32 位整数范围来说，可以视为 $O(1)$。
*   **总计：** 因为我们对 $n$ 个元素中的每一个都只进行了常数次操作，所以总时间复杂度是线性的 $O(n)$。

### 2. 空间复杂度 (Space Complexity)
**结论：$O(n)$**

*   **辅助空间：** 我们使用了一个 `stack` 来存储操作数。
*   **最坏情况：** 栈的大小取决于输入的结构。
    *   在最坏的情况下，表达式可能是一堆数字排在前面，操作符排在后面（例如：`["1", "2", "3", "4", "+", "+", "+"]`）。
    *   这种情况下，栈会一直增长，直到所有的数字都入栈。对于包含 $n$ 个元素的有效逆波兰表达式，栈中最多大约会存储 $n/2 + 1$ 个数字（即一半是数字，一半是操作符）。
*   **量级：** 因为栈的最大深度与输入规模 $n$ 成正比，所以空间复杂度为 $O(n)$。

### 总结

| 维度 | 复杂度 | 说明 |
| :--- | :--- | :--- |
| **时间** | **$O(n)$** | 需要遍历一次数组，每个元素处理耗时为常数。 |
| **空间** | **$O(n)$** | 需要一个栈来存储中间数字，栈的深度最多与输入长度成正比。 |

## 155. Min Stack

````python
"""
https://leetcode.com/problems/min-stack/description/

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the MinStack class:

MinStack() initializes the stack object.
void push(int val) pushes the element val onto the stack.
void pop() removes the element on the top of the stack.
int top() gets the top element of the stack.
int getMin() retrieves the minimum element in the stack.
You must implement a solution with O(1) time complexity for each function.

Example 1:

Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]

Explanation
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
 

Constraints:

-231 <= val <= 231 - 1
Methods pop, top and getMin operations will always be called on non-empty stacks.
At most 3 * 104 calls will be made to push, pop, top, and getMin.

"""

````

````python
class MinStack(object):

    def __init__(self):
        self.stack = []
        self.min_stack = []
        
    def push(self, val):
        """
        :type val: int
        :rtype: None
        """
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1] :
            self.min_stack.append(val)

    def pop(self):
        """
        :rtype: None
        """
        val = self.stack.pop()
        if val == self.min_stack[-1]:
            self.min_stack.pop()

    def top(self):
        """
        :rtype: int
        """
        return self.stack[-1]
        
    def getMin(self):
        """
        :rtype: int
        """
        return self.min_stack[-1]
        

# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
````

这是一个非常经典的面试题。要在 $O(1)$ 时间复杂度内实现 `getMin`，关键在于**用空间换时间**。

我们不能每次调用 `getMin` 时都去遍历整个栈，因此我们需要在 `push` 数据的时候，顺便把当前的最小值记录下来。

最通用的解法是使用 **辅助栈（Auxiliary Stack）**。

### 核心思路：双栈法

1.  **数据栈 (`stack`)**：正常存放所有压入的数据。
2.  **最小栈 (`min_stack`)**：存放**单调递减**的数据，栈顶永远是当前的最小值。

**逻辑细节：**

*   **Push(x)**:
    *   直接把 `x` 压入 `stack`。
    *   判断 `min_stack`：如果 `min_stack` 为空，或者 `x` **小于等于** `min_stack` 的栈顶元素，则将 `x` 也压入 `min_stack`。
    *   *注意：这里必须是小于等于（`<=`），以处理重复的最小值。*
*   **Pop()**:
    *   从 `stack` 弹出一个值 `val`。
    *   如果 `val` 等于 `min_stack` 的栈顶元素，说明这个最小值被移除了，那么 `min_stack` 也要同步弹出栈顶。
*   **Top()**:
    *   返回 `stack` 的栈顶。
*   **GetMin()**:
    *   返回 `min_stack` 的栈顶。

### Python 代码实现

```python
class MinStack:

    def __init__(self):
        # 主栈：存储所有元素
        self.stack = []
        # 辅助栈：存储当前的最小值序列
        self.min_stack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        # 如果辅助栈为空，或者新元素 <= 当前最小值，则压入辅助栈
        # 注意：这里必须包含等于的情况，防止重复最小值被误删
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self) -> None:
        # 弹出主栈元素
        val = self.stack.pop()
        # 如果弹出的元素就是当前的最小值，那么辅助栈也要弹出
        if val == self.min_stack[-1]:
            self.min_stack.pop()

    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        # 辅助栈的栈顶始终是当前主栈中的最小值
        return self.min_stack[-1]


# 使用示例:
# obj = MinStack()
# obj.push(-2)
# obj.push(0)
# obj.push(-3)
# param_4 = obj.getMin() # 返回 -3
# obj.pop()
# param_3 = obj.top()    # 返回 0
# param_4 = obj.getMin() # 返回 -2
```

### 复杂度分析

1.  **时间复杂度 (Time Complexity):**
    *   `push`: $O(1)$。主要是列表 append 操作。
    *   `pop`: $O(1)$。主要是列表 pop 操作。
    *   `top`: $O(1)$。索引访问。
    *   `getMin`: $O(1)$。直接访问 `min_stack` 的栈顶。
    *   **满足题目要求。**

2.  **空间复杂度 (Space Complexity):**
    *   $O(N)$。我们需要存储 $N$ 个元素。
    *   在最坏情况下（例如插入的数据是递减的：`5, 4, 3, 2, 1`），`min_stack` 的大小会和 `stack` 一样大。

### 图解演示

假设执行以下操作：
`push(-2)`, `push(0)`, `push(-3)`, `pop()`

1.  **push(-2)**:
    *   `stack`: `[-2]`
    *   `min_stack`: `[-2]` (因为为空，直接入)
2.  **push(0)**:
    *   `stack`: `[-2, 0]`
    *   `min_stack`: `[-2]` (0 > -2，不入 `min_stack`)
3.  **push(-3)**:
    *   `stack`: `[-2, 0, -3]`
    *   `min_stack`: `[-2, -3]` (-3 < -2，入 `min_stack`)
    *   *此时 getMin() 返回 -3*
4.  **pop()**:
    *   `stack` 弹出 `-3`。
    *   检查：弹出的 `-3` 等于 `min_stack` 栈顶 `-3`。
    *   `min_stack` 弹出 `-3`。
    *   `stack`: `[-2, 0]`
    *   `min_stack`: `[-2]`
    *   *此时 getMin() 返回 -2*

## 21. Merge Two Sorted Lists

````python
"""  
https://leetcode.com/problems/merge-two-sorted-lists/description/

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.



"""
````

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        dummy = ListNode()
        current = dummy

        while list1 and list2:
            if list1.val < list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next
        
        if list1:
            current.next = list1
        if list2:
            current.next = list2
        
        return dummy.next
        
````

这道题是 LeetCode 第 21 题 **Merge Two Sorted Lists**。

这是一个非常经典的链表题目。最通用的解法是使用 **“哨兵节点（Dummy Node）” + “迭代（Iterative）”** 的方式。

### 核心思路

想象你有两摞已经排好序的扑克牌（从小到大），你想把它们合成一摞有序的牌。
1.  你比较两摞牌顶部的牌。
2.  哪张小，就把它拿走放到新的一摞里。
3.  重复这个过程，直到其中一摞牌空了。
4.  把剩下一摞牌直接接在新牌堆的后面。

### Python 代码实现（迭代法 - 推荐）

这种方法空间复杂度最低，不容易爆栈。

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # 1. 创建一个哨兵节点（Dummy Node）
        # 它的值不重要，它的 next 将指向合并后的链表头
        dummy = ListNode(-1)
        
        # current 指针用来追踪新链表的尾部
        current = dummy
        
        # 2. 当两个链表都还有节点时，进行比较和拼接
        while list1 and list2:
            if list1.val <= list2.val:
                current.next = list1   # 把 list1 的节点接过来
                list1 = list1.next     # list1 指针后移
            else:
                current.next = list2   # 把 list2 的节点接过来
                list2 = list2.next     # list2 指针后移
            
            # current 指针永远指向新链表的最后一个节点
            current = current.next
        
        # 3. 处理剩余部分
        # 循环结束后，肯定有一个链表先空了，另一个链表可能还有剩余
        # 直接把剩余的链表接在 current 后面即可（因为它们本身就是有序的）
        if list1:
            current.next = list1
        elif list2:
            current.next = list2
            
        # 4. 返回哨兵节点的下一个节点（真正的头节点）
        return dummy.next
```

### 复杂度分析

*   **时间复杂度**: $O(N + M)$。其中 $N$ 和 $M$ 分别是两个链表的长度。我们需要遍历两个链表的每一个节点一次。
*   **空间复杂度**: $O(1)$。我们只是调整了现有节点的 `next` 指针，没有创建新的节点数据结构（除了一个哨兵节点）。

---

### 另一种写法（递归法 - 只有代码简洁）

虽然递归代码短，但在工程中如果链表非常长，可能会导致栈溢出（Stack Overflow），所以面试中优先写上面的迭代法。

```python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # 递归终止条件：如果其中一个为空，直接返回另一个
        if not list1:
            return list2
        if not list2:
            return list1
        
        # 递归逻辑
        if list1.val <= list2.val:
            # list1 小，那么 list1 就是头，它的 next 是 "list1剩余部分" 和 "list2" 合并的结果
            list1.next = self.mergeTwoLists(list1.next, list2)
            return list1
        else:
            # list2 小，同理
            list2.next = self.mergeTwoLists(list1, list2.next)
            return list2
```

### 总结
推荐掌握 **迭代法（Iterative）**，特别是 **Dummy Node（哨兵节点）** 的技巧，这在很多链表题目（如删除节点、反转链表）中都非常有用，可以避免处理头节点为空的麻烦逻辑。


## 141. Linked List Cycle

````python
""" 
https://leetcode.com/problems/linked-list-cycle/description/

Given head, the head of a linked list, determine if the linked list has a cycle in it.

There is a cycle in a linked list if there is some node in the list that 
can be reached again by continuously following the next pointer. 
Internally, pos is used to denote the index of the node that tail's next pointer is connected to. 
Note that pos is not passed as a parameter.

Return true if there is a cycle in the linked list. Otherwise, return false.

"""
````

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None


class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        seen = set()
        while head:
            if head in seen:
                return True
            seen.add(head)
            head = head.next
        return False
        
````

这道题是 LeetCode 第 141 题 **Linked List Cycle**。

判断链表是否有环，最经典、最高效的解法是 **“快慢指针法”**（也叫 Floyd 判圈算法，俗称“龟兔赛跑”）。

### 核心思路：龟兔赛跑

想象两个人在赛道上跑步：
1.  **乌龟（slow）**：一次跑 1 步。
2.  **兔子（fast）**：一次跑 2 步。

**情况 A：如果跑道是直的（没有环）**
兔子跑得快，一定会先到达终点（遇到 `None`），比赛结束。

**情况 B：如果跑道是圆形的（有环）**
兔子跑得快，它会一直在圈里转。迟早有一刻，兔子会“套圈”，从后面追上乌龟。**只要它们相遇了，就说明一定有环。**

### Python 代码实现

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        # 1. 特殊情况处理：如果是空链表或者只有一个节点且无环
        if not head or not head.next:
            return False
        
        # 2. 初始化指针
        slow = head
        fast = head
        
        # 3. 开始赛跑
        # 只有当 fast 和 fast.next 都不为空时，才能继续跑
        # 如果 fast 跑到 None 了，说明没有环
        while fast and fast.next:
            slow = slow.next        # 慢指针走一步
            fast = fast.next.next   # 快指针走两步
            
            # 4. 检查是否相遇
            if slow == fast:
                return True         # 追上了，说明有环
        
        # 5. 循环结束，说明 fast 跑到了尽头
        return False
```

### 复杂度分析

*   **时间复杂度**: $O(N)$。
    *   如果没有环，快指针跑完链表就需要 $N/2$ 次操作，即 $O(N)$。
    *   如果有环，快慢指针进入环后，快指针每次比慢指针多走 1 步，它们之间的距离每走一次缩小 1，因此在有限次循环内一定能相遇。
*   **空间复杂度**: $O(1)$。我们只用了两个指针（变量），没有使用额外的数组或哈希表。

---

### 另一种解法（哈希表法）

虽然不如快慢指针省空间，但这种逻辑更符合直觉。

**思路**：带一个小本子（Set），每到一个节点，就查一下本子上记过没。记过就是有环，没记过就写上去。

```python
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        seen = set()
        current = head
        while current:
            if current in seen:
                return True
            seen.add(current)
            current = current.next
        return False
```
*   **空间复杂度**: $O(N)$，因为需要存储所有节点。**面试时推荐优先写快慢指针法。**

## 206. Reverse Linked List

````python
"""  
https://leetcode.com/problems/reverse-linked-list/description/

Given the head of a singly linked list, reverse the list, and return the reversed list.
"""
````

````python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

        
class Solution(object):
    def reverseList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        prev = None
        cur = head
        while cur:
            temp = cur.next

            cur.next = prev
            prev = cur
            cur = temp
        
        return prev
        
````

这道题是 LeetCode 第 206 题 **Reverse Linked List**。

这也是链表中最基础、最常考的题目。核心解法有两种：**双指针迭代法**（推荐）和 **递归法**。

### 方法一：双指针迭代法（Iterative）—— 推荐

这是最通用、空间复杂度最低的方法。

#### 核心思路
我们需要三个指针（或者说是变量）来完成“调转方向”的操作：
1.  `curr`（当前节点）：指向我们正在处理的节点。
2.  `prev`（前一个节点）：指向 `curr` 后面那个节点（反转后，`curr` 要指向 `prev`）。
3.  `temp`（临时节点）：用来保存 `curr` 原本的下一个节点，防止断链。

**步骤：**
每次循环中，我们将 `curr` 的指针指向 `prev`，然后把 `prev` 和 `curr` 整体向后移动一步。

#### Python 代码

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None        # 初始化前驱节点为空（因为反转后的尾节点指向 None）
        curr = head        # 当前节点从头开始
        
        while curr:
            # 1. 暂存下一步要去的地方，防止断链
            temp = curr.next
            
            # 2. 【核心操作】将当前节点的指针反向，指向前一个节点
            curr.next = prev
            
            # 3. 两个指针整体向前移动
            prev = curr      # prev 走到当前位置
            curr = temp      # curr 走到下一格
            
        # 循环结束后，curr 变成了 None，而 prev 正好停在新的头节点上
        return prev
```

#### 复杂度分析
*   **时间复杂度**: $O(N)$，遍历了一次链表。
*   **空间复杂度**: $O(1)$，只使用了常数个变量。

### 方法二：递归法（Recursive）

递归法写起来代码非常短，很优雅，但是理解起来稍微绕一点。

#### 核心思路
假设链表是 `1 -> 2 -> 3 -> 4 -> 5`。
我们要反转它，可以看作：先反转 `2 -> ... -> 5` 这一段，得到 `5 -> 4 -> 3 -> 2`，然后让 `2` 指向 `1`，最后让 `1` 指向 `None`。

#### Python 代码

```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # 终止条件：如果是空链表，或者只剩一个节点，不需要反转，直接返回
        if not head or not head.next:
            return head
        
        # 递归调用：先去反转剩下的链表
        # new_head 将会一直是原本链表的最后一个节点（例如 5）
        new_head = self.reverseList(head.next)
        
        # 此时 head.next 指向的是链表里的下一个节点（比如当前 head 是 4，head.next 就是 5）
        # 我们要让 5 (head.next) 指向 4 (head)
        head.next.next = head
        
        # 断开 4 指向 5 的旧连接，防止形成环
        head.next = None
        
        return new_head
```

#### 复杂度分析
*   **时间复杂度**: $O(N)$。
*   **空间复杂度**: $O(N)$。因为递归调用需要消耗系统栈空间，深度为链表长度。

### 总结
面试中，**优先写方法一（迭代法）**，因为它不消耗额外的栈空间，不容易溢出。
