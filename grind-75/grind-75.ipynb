## 121. Best Time to Buy and Sell Stock

````python
"""
https://www.techinterviewhandbook.org/grind75/?grouping=topics

https://leetcode.com/problems/best-time-to-buy-and-sell-stock/description/

You are given an array prices where prices[i] is the price of a given stock on the ith day.
You want to maximize your profit by choosing a single day to buy one stock 
and choosing a different day in the future to sell that stock.
Return the maximum profit you can achieve from this transaction. 
If you cannot achieve any profit, return 0. 

Example 1:

Input: prices = [7,1,5,3,6,4]
Output: 5
Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.
Note that buying on day 2 and selling on day 1 is not allowed because you must buy before you sell.
Example 2:

Input: prices = [7,6,4,3,1]
Output: 0
Explanation: In this case, no transactions are done and the max profit = 0.
"""
````

````python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        max_profit = float('-inf')
        min_price = float('inf')
        for price in prices:
            min_price = min(min_price, price)
            max_profit = max(max_profit, price - min_price)
        return max_profit
````

这是一个非常经典的动态规划/贪心算法题目（LeetCode 121. Best Time to Buy and Sell Stock）。

### 解题思路：一次遍历 (One Pass)

我们不需要比较每一天和它之后所有日子的差值（那样是 $O(N^2)$）。我们只需要遍历一次数组，在遍历的过程中记录两个状态：
1.  **历史最低价格 (`min_price`)**：代表截止到当前为止，买入成本最低的那一天。
2.  **当前最大利润 (`max_profit`)**：代表如果在历史最低点买入，并在当前卖出，是否能打破之前的利润记录。

**逻辑步骤：**
*   初始化 `min_price` 为无穷大。
*   初始化 `max_profit` 为 0。
*   遍历每一天的价格 `price`：
    *   **情况 A**：如果 `price` 比 `min_price` 还低，说明我们要更新买入点（因为越低买入越好），更新 `min_price = price`。
    *   **情况 B**：如果 `price` 比 `min_price` 高，说明今天卖出能赚钱。计算 `price - min_price`，如果这个值比 `max_profit` 大，就更新 `max_profit`。

### Python 代码实现

```python
from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        # 如果数组为空或只有一个元素，无法进行交易，返回0
        if not prices:
            return 0
        
        # 初始化最低价格为正无穷大，最大利润为0
        min_price = float('inf')
        max_profit = 0
        
        for price in prices:
            # 1. 寻找历史最低价（假设这一天买入）
            if price < min_price:
                min_price = price
            
            # 2. 如果当前价格大于最低价，计算利润并更新最大值
            # 注意：这里的 elif 保证了我们不会在更新最低价的同一天卖出
            # (虽然同一天买卖利润为0，不影响结果，但逻辑上这样更清晰)
            elif price - min_price > max_profit:
                max_profit = price - min_price
                
        return max_profit

# --- 测试用例 ---
if __name__ == "__main__":
    sol = Solution()
    
    # 示例 1
    prices1 = [7, 1, 5, 3, 6, 4]
    print(f"Input: {prices1}")
    print(f"Output: {sol.maxProfit(prices1)}") 
    # 解释: 在第 2 天（价格=1）买入，在第 5 天（价格=6）卖出，利润 = 6-1 = 5。
    
    print("-" * 20)
    
    # 示例 2
    prices2 = [7, 6, 4, 3, 1]
    print(f"Input: {prices2}")
    print(f"Output: {sol.maxProfit(prices2)}") 
    # 解释: 价格持续下跌，找不到比买入价更高的卖出价，利润为 0。
```

### 复杂度分析

*   **时间复杂度**: $O(N)$
    *   只需要遍历 `prices` 数组一次。
*   **空间复杂度**: $O(1)$
    *   只使用了两个变量 `min_price` 和 `max_profit`。

### 图解推导 (示例 1: `[7, 1, 5, 3, 6, 4]`)

| 当前价格 (Price) | 动作判断 | 更新后的 min_price | 当前计算利润 | 更新后的 max_profit |
| :--- | :--- | :--- | :--- | :--- |
| **7** | 7 < inf (更新最低价) | **7** | - | 0 |
| **1** | 1 < 7 (发现更低价，更新) | **1** | - | 0 |
| **5** | 5 > 1 (计算利润) | 1 | 5 - 1 = 4 | **4** |
| **3** | 3 > 1 (计算利润) | 1 | 3 - 1 = 2 | 4 (不变) |
| **6** | 6 > 1 (计算利润) | 1 | 6 - 1 = 5 | **5** |
| **4** | 4 > 1 (计算利润) | 1 | 4 - 1 = 3 | 5 (不变) |

最终返回 **5**。
